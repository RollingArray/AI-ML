## Testing Of Hypothesis

### Q.
Manufacturer of product ‘A’ claims that average life of the product is 120 months with variance 25 months. A random sample of 10 of these is with mean 125 months and standard deviation 4 months. Validate the claim of the manufacturer at 1% level of significance. 

### A
To test the claim of the manufacturer, we can conduct a hypothesis test using the sample data and comparing it with the claimed population parameters. 

Let's set up the hypotheses:

Null Hypothesis ($H_0$): The average life of the product is indeed 120 months ($ \mu = 120 $).

Alternative Hypothesis ($H_1$): The average life of the product is not 120 months ($ \mu \neq 120 $).

We'll use a t-test for this hypothesis test because the population variance is unknown and we have a sample size less than 30. 

Given:
- Claimed population mean ($ \mu $): 120 months
- Claimed population variance ($ \sigma^2 $): 25 months (so the standard deviation, $ \sigma $, is $ \sqrt{25} = 5 $ months)
- Sample size ($ n $): 10
- Sample mean ($ \bar{x} $): 125 months
- Sample standard deviation ($ s $): 4 months

Let's calculate the t-value and compare it with the critical t-value at a 1% significance level (two-tailed test, since $H_1$ is not equal to $H_0$).

The formula for the t-value is:

$ t = \frac{\bar{x} - \mu}{\frac{s}{\sqrt{n}}} $

Let's plug in the values:

$ t = \frac{125 - 120}{\frac{4}{\sqrt{10}}} $

$ t = \frac{5}{\frac{4}{\sqrt{10}}} $

$ t = \frac{5}{\frac{4}{\sqrt{10}}} $

$ t \approx \frac{5}{1.264} $

$ t \approx 3.95 $

Now, we need to find the critical t-value at a 1% significance level with $ n - 1 = 9 $ degrees of freedom. 

Using a t-table or statistical software, the critical t-value is approximately $ \pm 3.250 $ for a two-tailed test at a 1% significance level.

Since our calculated t-value (3.95) is greater than the critical t-value (3.250), we reject the null hypothesis.

Therefore, we have enough evidence to conclude that the average life of the product is not 120 months at the 1% level of significance. This suggests that the manufacturer's claim might not be accurate.

### Question
It is claimed that player A is better than Player B in IPL matches during IPL bidding for players. During previous IPL season both played 18 matches with average runs scored 85 and 75 with standard deviations 9 and 2 respectively. Validate the claim using appropriate statistical procedure at 1% level of significance.    

### Solution
To test the claim that Player A is better than Player B in IPL matches, we can use a hypothesis test comparing their average runs scored. 

Let's set up the hypotheses:

Null Hypothesis ($H_0$): There is no difference between the average runs scored by Player A and Player B ($ \mu_A = \mu_B $).

Alternative Hypothesis ($H_1$): Player A scores more runs on average than Player B ($ \mu_A > \mu_B $).

We'll use a two-sample t-test for independent samples because we have two groups of data (runs scored by Player A and Player B) and we want to compare their means.

Given:
- Number of matches for both players ($ n $): 18
- Average runs scored by Player A ($ \bar{x}_A $): 85
- Average runs scored by Player B ($ \bar{x}_B $): 75
- Standard deviation of runs scored by Player A ($ \sigma_A $): 9
- Standard deviation of runs scored by Player B ($ \sigma_B $): 2

We'll calculate the t-value using the formula:

$ t = \frac{(\bar{x}_A - \bar{x}_B)}{\sqrt{\frac{s_A^2}{n} + \frac{s_B^2}{n}}} $

Where:
- $ \bar{x}_A $ and $ \bar{x}_B $ are the sample means of runs scored by Player A and Player B respectively.
- $ s_A $ and $ s_B $ are the sample standard deviations of runs scored by Player A and Player B respectively.
- $ n $ is the number of matches for each player.

Let's plug in the values:

$ t = \frac{(85 - 75)}{\sqrt{\frac{9^2}{18} + \frac{2^2}{18}}} $

$ t = \frac{10}{\sqrt{\frac{81}{18} + \frac{4}{18}}} $

$ t = \frac{10}{\sqrt{\frac{85}{18}}} $

$ t = \frac{10}{\sqrt{4.722}} $

$ t \approx \frac{10}{2.173} $

$ t \approx 4.60 $

Now, we need to find the critical t-value at a 1% significance level with $ n_1 + n_2 - 2 = 18 + 18 - 2 = 34 $ degrees of freedom.

Using a t-table or statistical software, the critical t-value for a one-tailed test at a 1% significance level with 34 degrees of freedom is approximately 2.750.

Since our calculated t-value (4.60) is greater than the critical t-value (2.750), we reject the null hypothesis.

Therefore, we have enough evidence to conclude that Player A is indeed better than Player B in terms of average runs scored at the 1% level of significance.

### Question
Formulate a suitable hypothesis and validate it by using appropriate statistical procedure based on the following data. (At 5% level of significance)
“Before pandemic 200 out of 750 students are attending online classes whereas after pandemic 350 out of 800 are doing the same”.   

### Solution
To test whether there has been a significant change in the proportion of students attending online classes before and after the pandemic, we can use the hypothesis test for comparing two proportions.

Let's set up the hypotheses:

Null Hypothesis ($H_0$): The proportion of students attending online classes before the pandemic is equal to the proportion after the pandemic ($p_1 = p_2$).

Alternative Hypothesis ($H_1$): The proportion of students attending online classes before the pandemic is not equal to the proportion after the pandemic ($p_1 \neq p_2$).

We'll use a two-sample z-test for proportions. 

Given:
- Before the pandemic: $n_1 = 750$, $x_1 = 200$ (students attending online classes)
- After the pandemic: $n_2 = 800$, $x_2 = 350$ (students attending online classes)

We need to calculate the sample proportions ($p_1$ and $p_2$), the combined proportion ($p$), and the standard error ($SE$).

$ p_1 = \frac{x_1}{n_1} = \frac{200}{750} $

$ p_2 = \frac{x_2}{n_2} = \frac{350}{800} $

$ p = \frac{x_1 + x_2}{n_1 + n_2} = \frac{200 + 350}{750 + 800} $

Once we have $p_1$, $p_2$, and $p$, we can calculate the standard error:

$ SE = \sqrt{p(1-p)\left(\frac{1}{n_1} + \frac{1}{n_2}\right)} $

Finally, we'll calculate the z-score:

$ z = \frac{p_1 - p_2}{SE} $

Let's perform the calculations:

$ p_1 = \frac{200}{750} \approx 0.2667 $

$ p_2 = \frac{350}{800} = 0.4375 $

$ p = \frac{200 + 350}{750 + 800} = \frac{550}{1550} \approx 0.3548 $

$ SE = \sqrt{0.3548 \times (1-0.3548)\left(\frac{1}{750} + \frac{1}{800}\right)} $

$ SE \approx \sqrt{0.3548 \times 0.6452 \times (0.001333 + 0.00125)} $

$ SE \approx \sqrt{0.2292 \times 0.002583} $

$ SE \approx \sqrt{0.00059184} $

$ SE \approx 0.0243 $

Now, let's calculate the z-score:

$ z = \frac{0.2667 - 0.4375}{0.0243} $

$ z \approx \frac{-0.1708}{0.0243} $

$ z \approx -7.027 $

Using a standard normal distribution table or statistical software, the critical z-value for a two-tailed test at a 5% significance level is approximately $\pm 1.96$.

Since our calculated z-value (-7.027) is less than the critical z-value, we reject the null hypothesis.

Therefore, we have enough evidence to conclude that there has been a significant change in the proportion of students attending online classes before and after the pandemic at the 5% level of significance.

### Q
Consider the following data related to education loan taken or not for pursuing higher education. Formulate a suitable hypothesis and validate it by using an appropriate statistical procedure.                                                                                                                                         
Education Loan	| Boys	| Girls
----------------|--------|------
Loan taken	| 300	| 100
Not taken	| 150	| 250

### A
To test whether there is a significant association between taking an education loan and gender, we can use the chi-square test for independence. This test assesses whether the frequency distribution of categorical variables is independent of each other.

Let's set up the hypotheses:

Null Hypothesis ($H_0$): There is no association between taking an education loan and gender.

Alternative Hypothesis ($H_1$): There is an association between taking an education loan and gender.

We'll use a significance level of 5%.

Given data:

$
\begin{array}{|c|c|c|c|}
\hline
\text{Education Loan} & \text{Boys} & \text{Girls} & \text{Total} \\
\hline
\text{Loan taken} & 300 & 100 & 400 \\
\text{Not taken} & 150 & 250 & 400 \\
\hline
\text{Total} & 450 & 350 & 800 \\
\hline
\end{array}
$

We need to calculate the expected frequencies for each cell under the assumption of independence. We can do this using the formula:

$ \text{Expected frequency} = \frac{\text{row total} \times \text{column total}}{\text{grand total}} $

Let's calculate the expected frequencies:

For the cell corresponding to "Loan taken" and "Boys":
$ E_{\text{Loan taken, Boys}} = \frac{400 \times 450}{800} = 225 $

For the cell corresponding to "Loan taken" and "Girls":
$ E_{\text{Loan taken, Girls}} = \frac{400 \times 350}{800} = 175 $

For the cell corresponding to "Not taken" and "Boys":
$ E_{\text{Not taken, Boys}} = \frac{400 \times 450}{800} = 225 $

For the cell corresponding to "Not taken" and "Girls":
$ E_{\text{Not taken, Girls}} = \frac{400 \times 350}{800} = 175 $

Now, we'll use the chi-square test statistic formula:

$ \chi^2 = \sum \frac{(O - E)^2}{E} $

Where $O$ is the observed frequency and $E$ is the expected frequency.

Let's calculate the chi-square value:

$ \chi^2 = \frac{(300 - 225)^2}{225} + \frac{(100 - 175)^2}{175} + \frac{(150 - 225)^2}{225} + \frac{(250 - 175)^2}{175} $

$ \chi^2 = \frac{25^2}{225} + \frac{75^2}{175} + \frac{75^2}{225} + \frac{75^2}{175} $

$ \chi^2 = \frac{625}{225} + \frac{5625}{175} + \frac{5625}{225} + \frac{5625}{175} $

$ \chi^2 \approx 2.78 + 32.14 + 25 + 32.14 $

$ \chi^2 \approx 91.06 $

Now, we need to find the critical chi-square value at a 5% significance level with degrees of freedom ($(r-1)(c-1)$), where $r$ is the number of rows and $c$ is the number of columns. In this case, $r = 2$ and $c = 2$, so the degrees of freedom is $df = (2-1)(2-1) = 1$.

Using a chi-square distribution table or statistical software, the critical chi-square value at a 5% significance level with 1 degree of freedom is approximately 3.84.

Since our calculated chi-square value (91.06) is greater than the critical chi-square value (3.84), we reject the null hypothesis.

Therefore, we have enough evidence to conclude that there is a significant association between taking an education loan and gender at the 5% level of significance.

### Q

Manufacturer of a drug claims that the one of their patented drugs is effective in curing pulmonary diseases with 75% efficiency. It is observed that 380 out of 500 patients used this drug found that it is effective in curing pulmonary diseases.  Validate the claim at 1% level of significance by using an appropriate test.                                                                                     

### A
To validate the claim made by the manufacturer regarding the efficiency of their drug, we can conduct a hypothesis test for a population proportion.

Let's set up the hypotheses:

Null Hypothesis ($H_0$): The proportion of patients cured by the drug is indeed 75% ($p = 0.75$).

Alternative Hypothesis ($H_1$): The proportion of patients cured by the drug is not 75% ($p \neq 0.75$).

We will use the z-test for population proportion.

Given:
- Number of patients who used the drug ($n$): 500
- Number of patients cured by the drug ($x$): 380
- Claimed population proportion ($p$): 0.75 (75%)

We need to calculate the sample proportion ($\hat{p}$) and the standard error ($SE$).

$ \hat{p} = \frac{x}{n} = \frac{380}{500} = 0.76 $

$ SE = \sqrt{\frac{p(1-p)}{n}} = \sqrt{\frac{0.75 \times (1-0.75)}{500}} $

$ SE = \sqrt{\frac{0.75 \times 0.25}{500}} $

$ SE = \sqrt{\frac{0.1875}{500}} $

$ SE = \sqrt{0.000375} $

$ SE \approx 0.0194 $

Now, we'll calculate the z-score:

$ z = \frac{\hat{p} - p}{SE} = \frac{0.76 - 0.75}{0.0194} $

$ z = \frac{0.01}{0.0194} $

$ z \approx 0.5155 $

Now, we need to find the critical z-value at a 1% significance level. For a two-tailed test, the critical z-value is approximately $ \pm 2.576 $.

Since our calculated z-value (0.5155) is less than the critical z-value (2.576), we fail to reject the null hypothesis.

Therefore, at the 1% level of significance, we do not have enough evidence to conclude that the proportion of patients cured by the drug is different from 75%. The claim made by the manufacturer is not contradicted by the data.

# What is testing of Hypothesis

## Analogy

Imagine you have a question in your mind about something. Let's say you're wondering if drinking a cup of coffee in the morning really helps you stay awake throughout the day. You have a hypothesis, which is just a fancy word for an educated guess or a statement that you think might be true. In this case, your hypothesis could be: "Drinking coffee in the morning helps me stay awake."

Now, you want to figure out if your hypothesis is true or not. Here's where testing comes in. You design a little experiment to test your hypothesis. You decide to have two groups: one group of people who drink coffee in the morning, and another group who don't drink any coffee at all. Then, you observe how awake they feel throughout the day.

After running your experiment and collecting data, you analyze the results. You compare the two groups to see if there's a noticeable difference in how awake they feel. If the group that drank coffee does indeed feel more awake on average compared to the group that didn't, it supports your hypothesis. But if there's no real difference between the two groups, then it suggests that your hypothesis might not be true.

This process of testing your hypothesis is what we call hypothesis testing. It's a way to scientifically investigate whether our ideas about the world are accurate or not. And it's not just about coffee – hypothesis testing is used in all kinds of fields, from medicine to economics to psychology, to help us make sense of the world around us.

## Mathematically
Hypothesis testing involves a series of steps:

1. **Formulate Hypotheses**: You start with two hypotheses:
   - **Null Hypothesis (H0)**: This is the default assumption, often stating that there is no effect or no difference. In our coffee example, the null hypothesis could be that drinking coffee in the morning does not help in staying awake.
   - **Alternative Hypothesis (H1 or Ha)**: This is what you're trying to find evidence for. It's typically the opposite of the null hypothesis. In our coffee example, the alternative hypothesis could be that drinking coffee in the morning does help in staying awake.

2. **Choose a Significance Level (α)**: This is the threshold for how much evidence you need to reject the null hypothesis. Common values for α are 0.05 or 0.01.

3. **Collect Data and Choose a Test Statistic**: You gather data from your experiment. Then, you choose a test statistic that helps you compare your observed data to what you would expect under the assumption that the null hypothesis is true. The choice of test statistic depends on the type of data you have and the specific hypothesis you're testing.

4. **Calculate the P-value**: This is the probability of observing your test statistic (or something more extreme) if the null hypothesis is true. A small p-value suggests that the observed data is unlikely to have occurred if the null hypothesis is true, which leads us to reject the null hypothesis in favor of the alternative hypothesis.

5. **Make a Decision**: If the p-value is less than the significance level (α), you reject the null hypothesis in favor of the alternative hypothesis. Otherwise, you fail to reject the null hypothesis.

6. **Draw Conclusions**: Based on your decision, you draw conclusions about whether there is enough evidence to support your alternative hypothesis.

In mathematical terms, hypothesis testing often involves calculating probabilities, using statistical distributions (like the normal distribution or t-distribution), and comparing observed data to theoretical expectations under the null hypothesis. It's a way to quantify and make decisions based on uncertainty in data analysis.

## Critical Region

Imagine you're conducting an experiment to test whether a new medicine helps people sleep better. You want to make sure your results are reliable and not just due to chance.

**Critical Region:**

1. **What it Represents:**
   - The critical region is like a special zone on a plot or distribution where unusual or unexpected results hang out.

2. **Why it Matters:**
   - In hypothesis testing, we compare our observed data or test statistic to a threshold called the critical value.
   - If our test statistic falls within this special zone (the critical region), it means our results are so unusual that they provide evidence against the null hypothesis (the default assumption that there's no difference between groups).

3. **How it's Determined:**
   - The critical region is determined by the chosen level of significance (α), which is usually set before conducting the experiment.
   - The level of significance represents the acceptable risk of making a Type I error (wrongly rejecting a true null hypothesis).
   - For example, if we choose α = 0.05, it means we're willing to accept a 5% chance of making a Type I error.

4. **Interpreting Results:**
   - If our test statistic falls within the critical region (to the right of the critical value), it means our observed results are unlikely to have occurred by random chance alone.
   - This suggests that there's a real effect or difference between groups, and we reject the null hypothesis in favor of an alternative hypothesis.

### Visualization using Python

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Generate data for a normal distribution
x = np.linspace(-4, 4, 1000)
y = stats.norm.pdf(x, loc=0, scale=1)

# Set significance level (alpha)
alpha = 0.05

# Calculate critical value based on alpha
critical_value = stats.norm.ppf(1 - alpha)

# Create plot
plt.figure(figsize=(10, 6))

# Plot the normal distribution curve
plt.plot(x, y, color='black', label='Normal Distribution')

# Shade the critical region
plt.fill_between(x[x > critical_value], stats.norm.pdf(x[x > critical_value], loc=0, scale=1), color='red', alpha=0.3, label='Critical Region')

# Annotate significance level and critical value
plt.axvline(critical_value, color='blue', linestyle='--', label='Critical Value')
plt.text(critical_value + 0.1, 0.1, f'α = {alpha}', fontsize=12, color='red')

# Add labels and title
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.title('Visualizing Critical Region')
plt.legend()
plt.grid(True)

# Show plot
plt.show()
```

In the plot:

1. **Normal Distribution Curve:**
   - The smooth curve represents a normal distribution, which is a common way to model real-world data. Each point on the curve shows the probability of a particular value occurring.

2. **Critical Region:**
   - The shaded area to the right of the critical value (indicated by the blue dashed line) represents the critical region.
   - This critical region is determined by the chosen significance level (α), which is usually set before conducting the experiment.
   - The significance level represents the probability of making a Type I error (wrongly rejecting a true null hypothesis). Here, α = 0.05 means we're willing to accept a 5% chance of making this error.

3. **Critical Value:**
   - The critical value is the boundary point that separates the critical region from the rest of the distribution.
   - It's determined based on the chosen significance level (α).
   - If our test statistic falls to the right of this critical value, it means our observed results are statistically significant at the chosen level of significance.

4. **Interpretation:**
   - The plot helps us understand how the critical region is defined based on the significance level. Anything beyond the critical value is considered unusual enough to provide evidence against the null hypothesis.
   - In hypothesis testing, if our test statistic falls within this critical region, it suggests that our observed results are unlikely to have occurred by random chance alone, leading us to reject the null hypothesis in favor of an alternative hypothesis.

## Level of Significance

Imagine you're investigating whether a new drug helps people sleep better. You want to make sure your results are reliable and not just due to chance.

The level of significance is like a threshold you set to decide how confident you want to be in your results.

1. **Setting the Threshold:**
   - You choose a level of significance before you start your experiment. It's typically denoted by the symbol α (alpha).
   - Common values for α are 0.05 (5%) or 0.01 (1%), but you can choose other values based on your needs.

2. **What it Means:**
   - If you set α to 0.05, it means you're willing to accept a 5% chance of making a Type I error. This error occurs when you wrongly reject a true null hypothesis.
   - In simpler terms, if you set α to 0.05, you're saying, "I'm okay with being wrong 5% of the time if I reject the null hypothesis when I shouldn't."

3. **Interpreting Results:**
   - After conducting your experiment, if your p-value (a measure of the strength of evidence against the null hypothesis) is less than α, you reject the null hypothesis.
   - If your p-value is greater than α, you don't reject the null hypothesis.
   - Essentially, the level of significance helps you decide how much evidence you need before you make a decision about the null hypothesis.

   - This critical region is determined by the chosen significance level (α), which is annotated on the plot.
   - In hypothesis testing, if our test statistic falls within this shaded area, it means the observed result is statistically significant at the chosen level of significance.

   - The critical value is the boundary point that separates the critical region from the rest of the distribution.
   - It's calculated based on the chosen significance level (α).

# P-value

Imagine you're testing a new medicine to see if it makes people feel better. You give the medicine to one group of people and a fake pill (placebo) to another group.

The p-value is like a score that tells you how likely it is that the medicine actually works. Here's how it works:

1. **What the P-Value Tells You:**
   - If the p-value is small (usually less than 0.05), it means that the difference you see between the group that took the medicine and the group that took the placebo is unlikely to be just a random chance.
   - In other words, a small p-value suggests that there's a real difference between the two groups, and the medicine might actually be making people feel better.

2. **How to Think About It:**
   - Think of the p-value as a measure of the strength of evidence against a null hypothesis. The null hypothesis is like the default assumption that there's no difference between the groups. 
   - A small p-value means that your data provides strong evidence against this null hypothesis, suggesting that there's something going on - maybe the medicine is really effective!

3. **What it Doesn't Tell You:**
   - A large p-value (greater than 0.05) doesn't mean that the medicine definitely doesn't work. It just means that the evidence isn't strong enough to confidently say that it does work.
   - It's like saying, "We can't be sure if the medicine works based on the data we have."

The p-value is a way to help you decide whether the results you're seeing are likely due to a real effect or just random chance. It's a measure of how confident you can be in your conclusions from the data you've collected.

### This table provides a quick reference guide for interpreting p-values in hypothesis testing:

| P-value   | Description                               | Hypothesis Interpretation |
|-----------|-------------------------------------------|---------------------------|
| ≤ 0.05    | Null hypothesis very unlikely             | Rejected                  |
| > 0.05    | Null hypothesis very likely               | Accepted or "fails to reject" |
| > 0.05    | P-value near the cut-off, considered marginal | Hypothesis needs more attention |

In essence:
- If the p-value is less than or equal to 0.05, it suggests strong evidence against the null hypothesis, so we reject it.
- If the p-value is greater than 0.05, it suggests weak evidence against the null hypothesis, so we accept it or "fail to reject" it.
- If the p-value is close to 0.05, it's considered marginal, indicating that more investigation may be needed to draw a conclusion.

# Errors in Decision Making

Imagine you're trying to decide whether to bring an umbrella with you when you go outside. You look at the sky and predict that it's going to rain based on the dark clouds you see. So, you decide to take the umbrella.

Now, there are two possible outcomes:

1. **It Rains (True Outcome)**:
   - You made the right decision to bring the umbrella because it actually rained. This is a correct decision.

2. **It Doesn't Rain (False Outcome)**:
   - However, it turns out that despite the dark clouds, it didn't rain at all. In this case, you made an incorrect decision by bringing the umbrella when it wasn't necessary.

These two outcomes represent the possibility of errors in decision making:

1. **False Positive (Type I Error)**:
   - This happens when you make a decision that suggests something is true when it's actually not. In our example, it would be bringing the umbrella (decision) when it doesn't rain (reality).

2. **False Negative (Type II Error)**:
   - This occurs when you fail to make a decision that would have been correct. In our example, it would be not bringing the umbrella (decision) when it does rain (reality).

### Mathematical
Type 1 Error: A Type-I error occurs when sample results reject the null hypothesis despite being true.

Type 2 Error: A Type-II error occurs when the null hypothesis is not rejected when it is false, unlike a Type-I error.

Example:
Suppose a teacher evaluates the examination paper to decide whether a student passes or fails.

H0: Student has passed

H1: Student has failed

Type I error will be the teacher failing the student [rejects H0] although the student scored the passing marks [H0 was true]. 

Type II error will be the case where the teacher passes the student [do not reject H0] although the student did not score the passing marks [H1 is true].

# Types of Hypothesis Testing

## One-Tail Test:

Imagine you're trying to find out if students who study with a particular method perform better on a test compared to those who don't use that method.

1. **What it Means:**
   - A one-tailed test, also known as a one-sided test, is like looking in just one direction for evidence.
   - It's like saying, "I'm only interested in knowing if one group performs better than the other, not worse."

2. **Choosing a Direction:**
   - Before you start your experiment, you decide which direction you want to look.
   - For example, you might want to know if students who use the new method score higher. That's a "right-tailed" one-tailed test.
   - Or you might want to know if students who use the new method score lower. That's a "left-tailed" one-tailed test.

3. **Interpreting the Results:**
   - After collecting data, you analyze it to see if the results support your hypothesis in the chosen direction.
   - If you're doing a right-tailed test and you find that the students who used the new method scored significantly higher than those who didn't, you'd conclude that the method is effective.
   - If you're doing a left-tailed test and you find that the students who used the new method scored significantly lower, you'd conclude that the method is not effective.

4. **When to Use:**
   - One-tailed tests are useful when you have a specific hypothesis about the direction of the effect.
   - They can be more powerful than two-tailed tests in detecting effects in one direction because they focus all the testing power on that direction.
  
Here's a mathematical table to explain one-tailed hypothesis testing:

| **Hypotheses**         | **Null Hypothesis (H0)**                                  | **Alternative Hypothesis (H1)**                      |
|------------------------|-----------------------------------------------------------|------------------------------------------------------|
| **Statement**          | The population parameter is equal to or less than some value | The population parameter is greater than that value  |
| **Symbolic Notation**  | $ H_0: \mu \leq \mu_0 $                                 | $ H_1: \mu > \mu_0 $                              |
| **Interpretation**     | No significant difference or less than specific value      | Greater than specific value                         |

In this table:

- **Hypotheses:** This column specifies the two hypotheses involved in the testing process.
- **Null Hypothesis (H0):** This hypothesis assumes that there is no significant difference or the population parameter is less than or equal to a specific value.
- **Alternative Hypothesis (H1):** This hypothesis suggests that there is a significant difference or the population parameter is greater than the specific value.
- **Statement:** It provides a verbal interpretation of each hypothesis.
- **Symbolic Notation:** It represents the hypotheses using mathematical symbols.
- **Interpretation:** It explains the implications of accepting or rejecting each hypothesis in terms of the population parameter's relationship to the specific value.

## Right-Tailed Test: 

1. **What it Means:**
   - In a right-tailed test, also known as an upper-tailed test, we're specifically interested in whether a particular outcome is greater than some expected value.
   - It's like looking to the right on a number line to see if the result is greater than what we'd expect by chance alone.

2. **Example:**
   - Suppose we're testing a new medication and we hypothesize that it increases patients' heart rates. A right-tailed test would determine if the medication leads to heart rates that are significantly higher than the baseline or compared to another group.

### Visual explanation using Python

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Generate sample data
np.random.seed(42)
sample_data = np.random.normal(loc=0, scale=1, size=100)

# Calculate sample mean and standard deviation
sample_mean = np.mean(sample_data)
sample_std = np.std(sample_data)

# Set significance level (alpha)
alpha = 0.05

# Calculate critical value based on alpha
critical_value = norm.ppf(1 - alpha)

# Calculate test statistic
test_statistic = (sample_mean - 0) / (sample_std / np.sqrt(len(sample_data)))

# Create plot
plt.figure(figsize=(10, 6))

# Plot normal distribution curve
x = np.linspace(-4, 4, 1000)
plt.plot(x, norm.pdf(x, loc=0, scale=1), color='black', label='Normal Distribution')

# Shade critical region
plt.fill_between(x[x > critical_value], norm.pdf(x[x > critical_value], loc=0, scale=1), color='red', alpha=0.3, label='Critical Region')

# Plot test statistic
plt.axvline(test_statistic, color='blue', linestyle='--', label='Test Statistic')

# Annotate significance level and critical value
plt.text(critical_value + 0.1, 0.1, f'α = {alpha}', fontsize=12, color='red')
plt.text(test_statistic + 0.1, 0.1, f'Test Statistic', fontsize=12, color='blue')

# Add labels and title
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.title('Right-Tailed Hypothesis Testing')
plt.legend()
plt.grid(True)

# Show plot
plt.show()
```

## Left-Tailed Test:

1. **What it Means:**
   - In a left-tailed test, also known as a lower-tailed test, we're specifically interested in whether a particular outcome is less than some expected value.
   - It's like looking to the left on a number line to see if the result is less than what we'd expect by chance alone.

2. **Example:**
   - Let's say we're testing a new teaching method, and we hypothesize that it decreases students' test anxiety. A left-tailed test would determine if students who undergo the new method experience significantly lower levels of test anxiety compared to those who don't.

### Visual explanation using Python

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Generate sample data
np.random.seed(42)
sample_data = np.random.normal(loc=0, scale=1, size=100)

# Calculate sample mean and standard deviation
sample_mean = np.mean(sample_data)
sample_std = np.std(sample_data)

# Set significance level (alpha)
alpha = 0.05

# Calculate critical value based on alpha
critical_value = norm.ppf(alpha)

# Calculate test statistic
test_statistic = (sample_mean - 0) / (sample_std / np.sqrt(len(sample_data)))

# Create plot
plt.figure(figsize=(10, 6))

# Plot normal distribution curve
x = np.linspace(-4, 4, 1000)
plt.plot(x, norm.pdf(x, loc=0, scale=1), color='black', label='Normal Distribution')

# Shade critical region
plt.fill_between(x[x < critical_value], norm.pdf(x[x < critical_value], loc=0, scale=1), color='red', alpha=0.3, label='Critical Region')

# Plot test statistic
plt.axvline(test_statistic, color='blue', linestyle='--', label='Test Statistic')

# Annotate significance level and critical value
plt.text(critical_value - 0.5, 0.1, f'α = {alpha}', fontsize=12, color='red')
plt.text(test_statistic - 0.5, 0.1, f'Test Statistic', fontsize=12, color='blue')

# Add labels and title
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.title('Left-Tailed Hypothesis Testing')
plt.legend()
plt.grid(True)

# Show plot
plt.show()
```

## Two-Tail Test:

Imagine you're trying to figure out if a new medicine affects people's heart rates. You want to know if it makes heart rates go up or down, but you're not sure which one it might be.

1. **What it Means:**
   - Two-tailed hypothesis testing is like saying, "Hey, let's see if this medicine does anything to heart rates, whether it makes them faster or slower."

2. **Setting Up Hypotheses:**
   - You start with a guess called the null hypothesis (H0). It's like saying, "The medicine doesn't do anything to heart rates. They're the same as usual."
   - Then, you have another guess called the alternative hypothesis (H1). It's like saying, "Actually, the medicine might make heart rates faster or slower. It's different from what we're used to."

3. **Example:**
   - Let's say you measure heart rates of people who took the medicine and those who didn't. You want to compare them.
   - The null hypothesis (H0) would be: "There's no difference in heart rates between those who took the medicine and those who didn't."
   - The alternative hypothesis (H1) would be: "There's a difference in heart rates between those who took the medicine and those who didn't."

4. **Interpreting the Results:**
   - If the test statistic is way out there in either direction (higher or lower), you might say, "Wow, that's weird. Our observed results are pretty different from what we expected if the medicine did nothing."
   - In that case, you'd reject the null hypothesis and say, "Hey, it looks like the medicine might be doing something to heart rates!"
   - But if the test statistic isn't too far out there, you might say, "Eh, our observed results aren't that unusual. Maybe the medicine doesn't really affect heart rates."
   - In that case, you'd fail to reject the null hypothesis and say, "We don't have enough evidence to say the medicine does anything to heart rates."

| **Hypotheses**         | **Null Hypothesis (H0)**                         | **Alternative Hypothesis (H1)**                           |
|------------------------|--------------------------------------------------|-----------------------------------------------------------|
| **Statement**          | The population parameter equals a specific value | The population parameter is not equal to that value        |
| **Symbolic Notation**  | $ H_0: \mu = \mu_0 $                           | $ H_1: \mu \neq \mu_0 $                                  |
| **Interpretation**     | No difference from the specific value            | Difference from the specific value (could be higher/lower) |

In this table:

- **Hypotheses:** This column specifies the two hypotheses involved in the testing process.
- **Null Hypothesis (H0):** This hypothesis assumes that there is no difference between the population parameter and a specific value.
- **Alternative Hypothesis (H1):** This hypothesis suggests that there is a difference between the population parameter and the specific value.
- **Statement:** It provides a verbal interpretation of each hypothesis.
- **Symbolic Notation:** It represents the hypotheses using mathematical symbols.
- **Interpretation:** It explains the implications of accepting or rejecting each hypothesis in terms of the population parameter's relationship to the specific value.

This table helps to understand the key components of two-tailed hypothesis testing and how to interpret the null and alternative hypotheses in mathematical terms.

### Visualization using Python

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Generate data for the plot
x = np.linspace(-5, 5, 1000)
y = norm.pdf(x, loc=0, scale=1)

# Set significance level (alpha)
alpha = 0.05

# Calculate critical values
critical_value_left = norm.ppf(alpha / 2)
critical_value_right = norm.ppf(1 - alpha / 2)

# Create plot
plt.figure(figsize=(10, 6))

# Plot normal distribution curve
plt.plot(x, y, color='black', label='Normal Distribution')

# Shade critical regions
plt.fill_between(x[x < critical_value_left], y[x < critical_value_left], color='red', alpha=0.3, label='Critical Region (Left)')
plt.fill_between(x[x > critical_value_right], y[x > critical_value_right], color='red', alpha=0.3, label='Critical Region (Right)')

# Plot critical values
plt.axvline(critical_value_left, color='blue', linestyle='--', label='Critical Value (Left)')
plt.axvline(critical_value_right, color='blue', linestyle='--', label='Critical Value (Right)')

# Add labels and title
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.title('Two-Tailed Hypothesis Testing')
plt.legend()
plt.grid(True)

# Show plot
plt.show()
```

In the plot

1. We generate data for the plot using a normal distribution centered at 0 with a standard deviation of 1.
2. We set the significance level (alpha) to 0.05.
3. We calculate the critical values for both tails of the distribution using the percent point function (ppf) of the normal distribution.
4. We create the plot, plotting the normal distribution curve and shading the critical regions in red.
5. We plot the critical values as dashed blue vertical lines.
6. Finally, we add labels, a title, a legend, and gridlines to the plot and display it.

## Z Test

Imagine you're trying to figure out if a new weight loss program actually helps people lose weight. You collect data from a group of people who follow the program and another group who don't. You want to compare their average weight loss to see if there's a significant difference.

**How does it work?**

| **Aspect**                  | **Explanation**                                                                 |
|-----------------------------|---------------------------------------------------------------------------------|
| **Purpose**                 | Determines if there's a significant difference between two groups or conditions |
| **Hypotheses**              | Null Hypothesis (H0): No difference between groups                              |
|                             | Alternative Hypothesis (H1): There is a difference between groups               |
| **Test Statistic**          | Z Score: Measures how many standard deviations a data point is from the mean     |
| **Critical Value**          | Obtained from a Z-table based on chosen significance level (usually 0.05)        |
| **Comparison**              | If Z Score falls into critical region, reject null hypothesis                    |
| **Conclusion**              | If null hypothesis rejected, conclude there's a significant difference           |

### Z-test formulas:

1. **One-Sample Z-test:**

   For testing a population mean when the population standard deviation (σ) is known:

   $ Z = \frac{\bar{x} - \mu}{\frac{\sigma}{\sqrt{n}}} $

   Where:
   - $ \bar{x} $ is the sample mean,
   - $ \mu $ is the population mean,
   - $ \sigma $ is the population standard deviation,
   - $ n $ is the sample size.

2. **Two-Sample Z-test:**

   For testing the difference between two population means when the population standard deviations (σ1 and σ2) are known:

   $ Z = \frac{$bar{x}_1 - \bar{x}_2) - $mu_1 - \mu_2)}{\sqrt{\frac{\sigma_1^2}{n_1} + \frac{\sigma_2^2}{n_2}}} $

   Where:
   - $ \bar{x}_1 $ and $ \bar{x}_2 $ are the sample means of the two groups,
   - $ \mu_1 $ and $ \mu_2 $ are the population means of the two groups,
   - $ \sigma_1 $ and $ \sigma_2 $ are the population standard deviations of the two groups,
   - $ n_1 $ and $ n_2 $ are the sample sizes of the two groups.

These formulas are used in hypothesis testing to calculate the Z-score, which is then compared to a critical value from a standard normal distribution to make a decision about the null hypothesis.

A taxi company is deciding whether to purchase brand A or brand B tires for its fleet of taxis. To estimate the difference in lifespan between the two brands, an experiment was conducted using 30 tires of brand A and 40 tires of brand B. The tires were run until they wore out. 

Test at a 5% level of significance if there is any significant difference between the two brands of tires.

- Brand A:
  - Sample size $(n_1$) = 30

  - Standard deviation $(SD_1$) = 5000 kilometers

### Solution

To decide whether to purchase brand A or brand B tires for the taxi fleet, we can conduct a hypothesis test to determine if there is a significant difference in the mean lifespan (in kilometers) between the two brands.

Given the data:

- Brand A:
  - Sample size $(n_1$) = 30
  - Mean $$bar{x}_1$) = 38600 kms
  - Standard Deviation $(SD_1$) = 5000 kms

- Brand B:
  - Sample size $(n_2$) = 40
  - Mean $$bar{x}_2$) = 35450 kms
  - Standard Deviation $(SD_2$) = 6100 kms

We'll perform a two-sample Z-test to compare the means of the two brands.

**Step 1: Formulate Hypotheses:**

- Null Hypothesis ($H_0$): There is no significant difference in the mean lifespan between brand A and brand B tires ($\mu_1 = \mu_2$).
- Alternative Hypothesis ($H_1$): There is a significant difference in the mean lifespan between brand A and brand B tires ($\mu_1 \neq \mu_2$).

**Step 2: Set Significance Level:**

We're given a significance level of 5%, so $alpha = 0.05$.

**Step 3: Calculate Test Statistic:**

$ Z = \frac{$bar{x}_1 - \bar{x}_2) - $mu_1 - \mu_2)}{\sqrt{\frac{SD_1^2}{n_1} + \frac{SD_2^2}{n_2}}} $

**Step 4: Determine Critical Value:**

For a two-tailed test at \$alpha = 0.05$, the critical value is approximately $Z = \pm 1.96$ (obtained from the Z-table).

**Step 5: Make Decision:**

If the absolute value of the test statistic is greater than 1.96, we reject the null hypothesis and conclude that there is a significant difference between the mean lifespans of the two brands.

**Calculation:**

$ Z = \frac{(38600 - 35450) - 0}{\sqrt{\frac{5000^2}{30} + \frac{6100^2}{40}}} $

Now, we calculate $Z$:

$ Z ≈ \frac{(3150) - 0}{\sqrt{\frac{25000000}{30} + \frac{37210000}{40}}} $

$ Z ≈ \frac{3150}{\sqrt{833333.33 + 930250}} $

$ Z ≈ \frac{3150}{\sqrt{1763083.33}} $

$ Z ≈ \frac{3150}{1328.97} $

$ Z ≈ 2.37 $

**Decision:**

Since $|Z| = 2.37 > 1.96$, we reject the null hypothesis. Therefore, there is a significant difference between the mean lifespans of brand A and brand B tires.

**Conclusion:**

Based on the test results, the taxi company should consider purchasing the brand of tires with the longer lifespan, which in this case appears to be brand A.

## Z-test for comparing two population proportions

Imagine you're comparing two groups, like students who studied with different methods, or patients who received different treatments. You want to know if one group had more success than the other, like getting higher grades or recovering faster.

Now, you collect some data from both groups. You count how many students passed in each group or how many patients recovered. But, because you're dealing with samples, not the entire population, you use proportions instead of raw numbers.

The formula you mentioned, the Z-test for comparing two proportions, helps you figure out if the difference you see in proportions between the two groups is likely due to a real difference or just random chance.

The formula you provided is for calculating the Z-test statistic for comparing two population proportions $(p_1 - p_2$). It is a variation of the Z-test formula for comparing proportions when the sample sizes are unequal.

Here's a breakdown of the formula:

$ Z = \frac{$hat{p}_1 - \hat{p}_2) - (p_1 - p_2)}{\sqrt{\frac{p \cdot q}{\left$frac{1}{n_1}\right) + \left$frac{1}{n_2}\right)}}} $

Where:
- $ \hat{p}_1 $ and $ \hat{p}_2 $ are the sample proportions of successes in sample 1 and sample 2, respectively.
- $ p_1 $ and $ p_2 $ are the hypothesized population proportions.
- $ p $ is the pooled proportion, calculated as $ p = \frac{x_1 + x_2}{n_1 + n_2} $, where $ x_1 $ and $ x_2 $ are the total number of successes in each sample, and $ n_1 $ and $ n_2 $ are the sample sizes.
- $ q $ is the complement of $ p $, calculated as $ q = 1 - p $.
- $ n_1 $ and $ n_2 $ are the sample sizes of sample 1 and sample 2, respectively.

This formula is used to calculate the Z-test statistic when the sample sizes are unequal. It takes into account the variability in the population proportions and adjusts for the different sample sizes.

### Example Problem
A pharmaceutical company is testing two different formulations of a new medication to determine their effectiveness in treating a certain medical condition. The company wants to know if there is a significant difference in the proportion of patients who experience symptom relief between the two formulations.

To conduct the study, the company randomly assigns 200 patients to receive formulation A and 250 patients to receive formulation B. After a specified treatment period, the company records the number of patients in each group who report symptom relief.

The data obtained from the study are as follows:

- Formulation A: Out of 200 patients, 120 report symptom relief.
- Formulation B: Out of 250 patients, 150 report symptom relief.

Using a significance level of 0.05, determine whether there is a statistically significant difference in the proportion of patients experiencing symptom relief between the two formulations.

Formulate appropriate hypotheses, calculate the Z-test statistic, and interpret the results to provide valuable insights to the pharmaceutical company regarding the efficacy of the two formulations in treating the medical condition.

### solution
To solve this problem, we will follow these steps:

1. Formulate hypotheses.
2. Calculate the proportions and the pooled proportion.
3. Calculate the standard error.
4. Calculate the Z-test statistic.
5. Determine the critical value and make a decision.
6. Interpret the results.

Let's proceed with the calculations:

1. **Formulate hypotheses:**
   - Null Hypothesis $(H_0$): There is no significant difference in the proportion of patients experiencing symptom relief between formulation A and formulation B. $(p_A = p_B$)
   - Alternative Hypothesis $(H_1$): There is a significant difference in the proportion of patients experiencing symptom relief between formulation A and formulation B. $(p_A \neq p_B$)

2. **Calculate proportions and the pooled proportion:**
   - Proportion for formulation A $(p_A$) = \$frac{120}{200} = 0.6$
   - Proportion for formulation B $(p_B$) = \$frac{150}{250} = 0.6$
   - Pooled proportion $(p$) = \$frac{120 + 150}{200 + 250} = \frac{270}{450} = 0.6$

3. **Calculate the standard error:**
   $ SE = \sqrt{\frac{p \cdot q}{\left$frac{1}{n_A}\right) + \left$frac{1}{n_B}\right)}} $
   $ SE = \sqrt{\frac{0.6 \cdot (1-0.6)}{\left$frac{1}{200}\right) + \left$frac{1}{250}\right)}} $
   $ SE = \sqrt{\frac{0.6 \cdot 0.4}{0.005 + 0.004}} $
   $ SE ≈ \sqrt{\frac{0.24}{0.009}} $
   $ SE ≈ \sqrt{26.67} $
   $ SE ≈ 5.16 $

4. **Calculate the Z-test statistic:**
   $ Z = \frac{(p_A - p_B)}{SE} $
   $ Z = \frac{0.6 - 0.6}{5.16} $
   $ Z = \frac{0}{5.16} $
   $ Z = 0 $

5. **Determine the critical value and make a decision:**
   Since this is a two-tailed test, we will compare the absolute value of Z with the critical value. With a significance level of 0.05, the critical Z-value is approximately ±1.96.

   Since \(|Z| = 0 < 1.96$, we fail to reject the null hypothesis.

6. **Interpret the results:**
   There is insufficient evidence to conclude that there is a significant difference in the proportion of patients experiencing symptom relief between formulation A and formulation B. Thus, the pharmaceutical company cannot conclude that one formulation is more effective than the other based on the data collected.

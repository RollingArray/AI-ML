# Principal Component Analysis (PCA)

Principal Component Analysis (PCA) is a dimensionality reduction technique commonly used in data science and machine learning. It works by finding the directions (or principal components) in which the data varies the most. These directions represent new axes in the feature space, ordered by the amount of variance they explain. By projecting the data onto these axes, PCA can effectively reduce the dimensionality while preserving most of the important information.

   ![Principal Component Analysis (PCA)](images/PCA.png)


Let's visually explain how PCA works step by step:

1. **Original Data:**
   Imagine you have a dataset with two features (variables), represented as points in a two-dimensional space. Each point corresponds to an observation in the dataset.

2. **Centering the Data:**
   PCA first centers the data by subtracting the mean of each feature from all data points. This ensures that the data is centered around the origin.

3. **Finding the First Principal Component (PC1):**
   PCA then finds the direction of maximum variance in the data. This direction is the first principal component (PC1). It's the line that best fits the data.

4. **Projecting Data onto PC1:**
   Next, PCA projects the data points onto PC1. This means dropping perpendicular lines from each point to PC1. The projected points are the data's new coordinates along PC1.

5. **Finding the Second Principal Component (PC2):**
   PCA then finds the second principal component (PC2), which is orthogonal (perpendicular) to PC1 and captures the remaining variance in the data.

6. **Projecting Data onto PC2:**
   Similar to PC1, PCA projects the data points onto PC2, obtaining new coordinates along PC2.

7. **Reduced-Dimension Representation:**
   The original data, represented by two features, has been effectively transformed into a new coordinate system defined by the principal components. The dimensionality is reduced, yet most of the important information is retained.

By choosing fewer principal components (i.e., retaining only the top few components that capture most of the variance), PCA allows for efficient dimensionality reduction, making it easier to visualize and analyze high-dimensional datasets.

**Example Dataset**:
Suppose we have a dataset representing the heights and weights of individuals. We'll consider a small dataset with 5 individuals, and each individual is represented by two features: height (in inches) and weight (in pounds).

| Individual | Height (inches) | Weight (pounds) |
|------------|-----------------|-----------------|
| 1          | 65              | 130             |
| 2          | 68              | 160             |
| 3          | 63              | 125             |
| 4          | 70              | 180             |
| 5          | 66              | 140             |

**Steps to Solve PCA Analytically**:

1. **Center the Data**:
   Calculate the mean of each feature (height and weight) and subtract it from each data point to center the data.

2. **Compute the Covariance Matrix**:
   Compute the covariance matrix of the centered data. The covariance matrix summarizes the relationships between all pairs of features.

3. **Find Eigenvalues and Eigenvectors**:
   Solve the characteristic equation $ \text{det}(A - \lambda I) = 0 $ to find the eigenvalues ($ \lambda $) and then solve $ (A - \lambda I) \mathbf{v} = \mathbf{0} $ to find the corresponding eigenvectors ($ \mathbf{v} $).

4. **Select Principal Components**:
   Select the principal components based on the highest eigenvalues. In this example, since we have two features (height and weight), we will have two principal components.

5. **Project Data onto Principal Components**:
   Project the original data onto the selected principal components to obtain the reduced-dimensional representation.

Certainly! Let's go through each step fully to solve PCA for the given example dataset.

### Step 1: Center the Data

Calculate the mean of each feature and subtract it from each data point to center the data.

Mean Height ($ \text{Mean}(h) $):
$ \text{Mean}(h) = \frac{65 + 68 + 63 + 70 + 66}{5} = 66.4 $

Mean Weight ($ \text{Mean}(w) $):
$ \text{Mean}(w) = \frac{130 + 160 + 125 + 180 + 140}{5} = 147 $

Subtract the means from the data points:
$ 
\begin{align*}
h' &= [65 - 66.4, 68 - 66.4, 63 - 66.4, 70 - 66.4, 66 - 66.4] \\
&= [-1.4, 1.6, -3.4, 3.6, -0.4] \\
w' &= [130 - 147, 160 - 147, 125 - 147, 180 - 147, 140 - 147] \\
&= [-17, 13, -22, 33, -7]
\end{align*}
$

### Step 2: Compute the Covariance Matrix

Compute the covariance between height and weight, as well as the variances of each feature:

Covariance between Height and Weight ($ \text{Cov}(h', w') $):
$ \text{Cov}(h', w') = \frac{1}{5 - 1} \sum_{i=1}^{5} (h_i' - \text{Mean}(h'))(w_i' - \text{Mean}(w')) $
$ \text{Cov}(h', w') = \frac{1}{4} [(-1.4 \times -17) + (1.6 \times 13) + (-3.4 \times -22) + (3.6 \times 33) + (-0.4 \times -7)] $
$ \text{Cov}(h', w') = \frac{1}{4} [-23.8 + 20.8 - 74.8 + 118.8 + 2.8] $
$ \text{Cov}(h', w') = \frac{1}{4} \times 43.8 = 10.95 $

Variance of Height ($ \text{Var}(h') $):
$ \text{Var}(h') = \frac{1}{5 - 1} \sum_{i=1}^{5} (h_i' - \text{Mean}(h'))^2 $
$ \text{Var}(h') = \frac{1}{4} [(-1.4)^2 + (1.6)^2 + (-3.4)^2 + (3.6)^2 + (-0.4)^2] $
$ \text{Var}(h') = \frac{1}{4} [1.96 + 2.56 + 11.56 + 12.96 + 0.16] $
$ \text{Var}(h') = \frac{1}{4} \times 29.2 = 7.3 $

Variance of Weight ($ \text{Var}(w') $):
$ \text{Var}(w') = \frac{1}{5 - 1} \sum_{i=1}^{5} (w_i' - \text{Mean}(w'))^2 $
$ \text{Var}(w') = \frac{1}{4} [(-17)^2 + 13^2 + (-22)^2 + 33^2 + (-7)^2] $
$ \text{Var}(w') = \frac{1}{4} [289 + 169 + 484 + 1089 + 49] $
$ \text{Var}(w') = \frac{1}{4} \times 2080 = 520 $

The covariance matrix:
$ 
\Sigma = 
\begin{bmatrix}
\text{Var}(h') & \text{Cov}(h', w') \\
\text{Cov}(h', w') & \text{Var}(w')
\end{bmatrix}
=
\begin{bmatrix}
7.3 & 10.95 \\
10.95 & 520
\end{bmatrix}
$


Sure, let's continue with Step 3 to find the eigenvalues and eigenvectors.

### Step 3: Find Eigenvalues and Eigenvectors

We already computed the covariance matrix $ \Sigma $:

$ 
\Sigma = 
\begin{bmatrix}
7.3 & 10.95 \\
10.95 & 520
\end{bmatrix}
$

Now, we need to find the eigenvalues ($ \lambda $) by solving the characteristic equation $ \text{det}(\Sigma - \lambda I) = 0 $.

The characteristic equation is given by:

$ \text{det}(\Sigma - \lambda I) = \text{det}\left(
\begin{bmatrix}
7.3 - \lambda & 10.95 \\
10.95 & 520 - \lambda
\end{bmatrix}\right) = 0 $

Expanding the determinant, we get:

$ (7.3 - \lambda)(520 - \lambda) - (10.95)^2 = 0 $

$ (7.3)(520) - 7.3\lambda - 520\lambda + \lambda^2 - (10.95)^2 = 0 $

$ \lambda^2 - (7.3 + 520)\lambda + (7.3 \times 520 - 10.95^2) = 0 $

$ \lambda^2 - 527.3\lambda + 3807.49 - 1202.5 = 0 $

$ \lambda^2 - 527.3\lambda + 2604.99 = 0 $

Now, we solve this quadratic equation to find the eigenvalues ($ \lambda $). We can use the quadratic formula:

$ \lambda = \frac{-b \pm \sqrt{b^2 - 4ac}}{2a} $

where $ a = 1 $, $ b = -527.3 $, and $ c = 2604.99 $.

$ \lambda = \frac{-(-527.3) \pm \sqrt{(-527.3)^2 - 4(1)(2604.99)}}{2(1)} $

$ \lambda = \frac{527.3 \pm \sqrt{278438.29 - 10419.96}}{2} $

$ \lambda = \frac{527.3 \pm \sqrt{268018.33}}{2} $

$ \lambda = \frac{527.3 \pm 517.71}{2} $

$ \lambda_1 \approx \frac{527.3 + 517.71}{2} \approx 522.505 $

$ \lambda_2 \approx \frac{527.3 - 517.71}{2} \approx 4.795 $

So, the eigenvalues are approximately $ \lambda_1 \approx 522.505 $ and $ \lambda_2 \approx 4.795 $.

Now, for each eigenvalue, we find the corresponding eigenvector by solving the equation $ (\Sigma - \lambda I) \mathbf{v} = \mathbf{0} $.

### For $ \lambda_1 \approx 522.505 $:

$ (\Sigma - \lambda_1 I) \mathbf{v}_1 = \mathbf{0} $

$ 
\left(
\begin{bmatrix}
7.3 & 10.95 \\
10.95 & 520
\end{bmatrix}
-
\begin{bmatrix}
522.505 & 0 \\
0 & 522.505
\end{bmatrix}
\right)
\mathbf{v}_1 = \mathbf{0}
$

$ 
\begin{bmatrix}
-515.205 & 10.95 \\
10.95 & -2.505
\end{bmatrix}
\mathbf{v}_1 = \mathbf{0}
$

Solving this system of equations will give us the eigenvector $ \mathbf{v}_1 $.

### For $ \lambda_2 \approx 4.795 $:

$ (\Sigma - \lambda_2 I) \mathbf{v}_2 = \mathbf{0} $

$ 
\left(
\begin{bmatrix}
7.3 & 10.95 \\
10.95 & 520
\end{bmatrix}
-
\begin{bmatrix}
4.795 & 0 \\
0 & 4.795
\end{bmatrix}
\right)
\mathbf{v}_2 = \mathbf{0}
$

$ 
\begin{bmatrix}
2.505 & 10.95 \\
10.95 & 515.205
\end{bmatrix}
\mathbf{v}_2 = \mathbf{0}
$

Solving this system of equations will give us the eigenvector $ \mathbf{v}_2 $.

Sure, let's continue with finding the eigenvectors.

### For $ \lambda_1 \approx 522.505 $:

$ 
\begin{bmatrix}
-515.205 & 10.95 \\
10.95 & -2.505
\end{bmatrix}
\mathbf{v}_1 = \mathbf{0}
$

We solve this system of equations by setting up the equations:

$ -515.205 v_{1_1} + 10.95 v_{1_2} = 0 $
$ 10.95 v_{1_1} - 2.505 v_{1_2} = 0 $

From the second equation, we can express $ v_{1_1} $ in terms of $ v_{1_2} $:
$ v_{1_1} = \frac{2.505}{10.95} v_{1_2} $

Substituting this expression into the first equation:
$ -515.205 \left(\frac{2.505}{10.95}\right) v_{1_2} + 10.95 v_{1_2} = 0 $

$ -118.725 v_{1_2} + 10.95 v_{1_2} = 0 $

$ -107.775 v_{1_2} = 0 $

This implies that $ v_{1_2} = 0 $.

So, $ v_{1_1} = \frac{2.505}{10.95} \times 0 = 0 $.

Therefore, the eigenvector corresponding to $ \lambda_1 \approx 522.505 $ is $ \mathbf{v}_1 = [0, 1]^T $.

### For $ \lambda_2 \approx 4.795 $:

$ 
\begin{bmatrix}
2.505 & 10.95 \\
10.95 & 515.205
\end{bmatrix}
\mathbf{v}_2 = \mathbf{0}
$

We solve this system of equations by setting up the equations:

$ 2.505 v_{2_1} + 10.95 v_{2_2} = 0 $
$ 10.95 v_{2_1} + 515.205 v_{2_2} = 0 $

From the first equation, we can express $ v_{2_1} $ in terms of $ v_{2_2} $:
$ v_{2_1} = -\frac{10.95}{2.505} v_{2_2} $

Substituting this expression into the second equation:
$ 10.95 \left(-\frac{10.95}{2.505}\right) v_{2_2} + 515.205 v_{2_2} = 0 $

$ -120.125 v_{2_2} + 515.205 v_{2_2} = 0 $

$ 395.08 v_{2_2} = 0 $

This implies that $ v_{2_2} = 0 $.

So, $ v_{2_1} = -\frac{10.95}{2.505} \times 0 = 0 $.

Therefore, the eigenvector corresponding to $ \lambda_2 \approx 4.795 $ is $ \mathbf{v}_2 = [1, 0]^T $.

So, we have found the eigenvectors corresponding to the eigenvalues $ \lambda_1 \approx 522.505 $ and $ \lambda_2 \approx 4.795 $.

In Step 4, we select the principal components based on the eigenvectors corresponding to the highest eigenvalues.

For our example, we found two eigenvalues:

1. $ \lambda_1 \approx 522.505 $
2. $ \lambda_2 \approx 4.795 $

The corresponding eigenvectors are:

1. $ \mathbf{v}_1 = [0, 1]^T $
2. $ \mathbf{v}_2 = [1, 0]^T $

Now, since $ \lambda_1 $ is much larger than $ \lambda_2 $, we choose the eigenvector corresponding to $ \lambda_1 $ as the first principal component, and the eigenvector corresponding to $ \lambda_2 $ as the second principal component.

So, our principal components are:

1. Principal Component 1: $ \mathbf{v}_1 = [0, 1]^T $
2. Principal Component 2: $ \mathbf{v}_2 = [1, 0]^T $

These principal components capture the directions of maximum variance in the data. The first principal component ($ \mathbf{v}_1 $) corresponds to the direction in which the data varies the most, while the second principal component ($ \mathbf{v}_2 $) captures the remaining variance orthogonal to the first component.

In Step 5, we project the original data onto the selected principal components to obtain the reduced-dimensional representation.

Let's denote the original centered data matrix as $ X' $, where each row represents an individual data point and each column represents a feature. In our case, $ X' $ will be a $ 5 \times 2 $ matrix.

$ 
X' = 
\begin{bmatrix}
-1.4 & -17 \\
1.6 & 13 \\
-3.4 & -22 \\
3.6 & 33 \\
-0.4 & -7
\end{bmatrix}
$

And let's denote the principal components matrix as $ V $, where each column represents a principal component. In our case, $ V $ will be a $ 2 \times 2 $ matrix.

$ 
V = 
\begin{bmatrix}
0 & 1 \\
1 & 0
\end{bmatrix}
$

To project the data onto the principal components, we simply multiply the data matrix $ X' $ by the principal components matrix $ V $:

$ X_{\text{projected}} = X' \times V $

$ 
X_{\text{projected}} = 
\begin{bmatrix}
-1.4 & -17 \\
1.6 & 13 \\
-3.4 & -22 \\
3.6 & 33 \\
-0.4 & -7
\end{bmatrix}
\times
\begin{bmatrix}
0 & 1 \\
1 & 0
\end{bmatrix}
$

$ 
X_{\text{projected}} = 
\begin{bmatrix}
-1.4(0) + (-17)(1) & -1.4(1) + (-17)(0) \\
1.6(0) + 13(1) & 1.6(1) + 13(0) \\
-3.4(0) + (-22)(1) & -3.4(1) + (-22)(0) \\
3.6(0) + 33(1) & 3.6(1) + 33(0) \\
-0.4(0) + (-7)(1) & -0.4(1) + (-7)(0)
\end{bmatrix}
$

$ 
X_{\text{projected}} = 
\begin{bmatrix}
-17 & -1.4 \\
13 & 1.6 \\
-22 & -3.4 \\
33 & 3.6 \\
-7 & -0.4
\end{bmatrix}
$

So, the projected data onto the principal components will be:

$ 
X_{\text{projected}} = 
\begin{bmatrix}
-17 & -1.4 \\
13 & 1.6 \\
-22 & -3.4 \\
33 & 3.6 \\
-7 & -0.4
\end{bmatrix}
$

This represents the reduced-dimensional representation of the original data using the principal components. Each row in this matrix corresponds to a data point, and each column corresponds to a principal component.

```python
import numpy as np

# Original centered data matrix
X = np.array([
    [-1.4, -17],
    [1.6, 13],
    [-3.4, -22],
    [3.6, 33],
    [-0.4, -7]
])

# Compute covariance matrix
covariance_matrix = np.cov(X, rowvar=False)

# Find eigenvalues and eigenvectors
eigenvalues, eigenvectors = np.linalg.eig(covariance_matrix)

# Sort eigenvalues and corresponding eigenvectors
sorted_indices = np.argsort(eigenvalues)[::-1]
eigenvalues = eigenvalues[sorted_indices]
eigenvectors = eigenvectors[:, sorted_indices]

# Select principal components
principal_components = eigenvectors[:, :2]

# Project data onto principal components
X_projected = np.dot(X, principal_components)

print("Original Data:")
print(X)
print("\nProjected Data:")
print(X_projected)
```

Let's perform PCA on this dataset and visualize the original and projected data points. The steps will be similar to the previous example, but with more features.

This code will plot the original data points in blue and the projected data points in red on the 3D scatter plot. This visualization will help us understand how PCA has reduced the dimensionality of the data while retaining the most important information.

```python
import numpy as np
import matplotlib.pyplot as plt

# Original data
X = np.array([
    [2104, 3, 5, 1.2, 10.6, 399.9],
    [1600, 3, 12, 0.8, 8.2, 329.9],
    [2400, 3, 8, 1.5, 12.5, 369.0],
    [1416, 2, 2, 1.0, 9.0, 232.0],
    [3000, 4, 10, 1.3, 14.0, 539.9],
    [1985, 4, 15, 0.9, 11.0, 299.9],
    [1534, 3, 6, 1.1, 10.2, 314.9],
    [1427, 3, 3, 0.7, 8.5, 198.9],
    [1380, 3, 4, 0.8, 9.3, 212.0],
    [1494, 3, 7, 1.0, 10.0, 242.0]
])

# Center the data
X_centered = X - np.mean(X, axis=0)

# Compute covariance matrix
covariance_matrix = np.cov(X_centered, rowvar=False)

# Find eigenvalues and eigenvectors
eigenvalues, eigenvectors = np.linalg.eig(covariance_matrix)

# Sort eigenvalues and corresponding eigenvectors
sorted_indices = np.argsort(eigenvalues)[::-1]
eigenvalues = eigenvalues[sorted_indices]
eigenvectors = eigenvectors[:, sorted_indices]

# Select principal components
num_principal_components = 2
principal_components = eigenvectors[:, :num_principal_components]

# Project data onto principal components
X_projected = np.dot(X_centered, principal_components)

# Plot original data
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X[:, 0], X[:, 1], X[:, 2], color='blue', label='Original Data')

# Plot projected data
ax.scatter(X_projected[:, 0], X_projected[:, 1], np.zeros_like(X_projected[:, 1]), color='red', label='Projected Data')

ax.set_xlabel('Size (sqft)')
ax.set_ylabel('Bedrooms')
ax.set_zlabel('Age (years)')
ax.set_title('PCA - Original vs Projected Data')
ax.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Original data
X = np.array([
    [2104, 3, 5, 1.2, 10.6, 399.9],
    [1600, 3, 12, 0.8, 8.2, 329.9],
    [2400, 3, 8, 1.5, 12.5, 369.0],
    [1416, 2, 2, 1.0, 9.0, 232.0],
    [3000, 4, 10, 1.3, 14.0, 539.9],
    [1985, 4, 15, 0.9, 11.0, 299.9],
    [1534, 3, 6, 1.1, 10.2, 314.9],
    [1427, 3, 3, 0.7, 8.5, 198.9],
    [1380, 3, 4, 0.8, 9.3, 212.0],
    [1494, 3, 7, 1.0, 10.0, 242.0]
])

# Center the data
X_centered = X - np.mean(X, axis=0)

# Compute covariance matrix
covariance_matrix = np.cov(X_centered, rowvar=False)

# Find eigenvalues and eigenvectors
eigenvalues, eigenvectors = np.linalg.eig(covariance_matrix)

# Sort eigenvalues and corresponding eigenvectors
sorted_indices = np.argsort(eigenvalues)[::-1]
eigenvalues = eigenvalues[sorted_indices]
eigenvectors = eigenvectors[:, sorted_indices]

# Select principal components
num_principal_components = 1 # Change to 1 or 2 for 1D or 2D projection, respectively
principal_components = eigenvectors[:, :num_principal_components]

# Project data onto principal components
X_projected = np.dot(X_centered, principal_components)

# Plot projected data
plt.scatter(X_projected[:, 0], np.zeros_like(X_projected[:, 0]), color='red', label='Projected Data')

plt.xlabel('Principal Component 1')
plt.title('PCA - Projected Data (1D)')
plt.legend()
plt.grid(True)
plt.show()
```

The output of the code will be a scatter plot representing the reduced-dimensional data after performing PCA. Here's how to interpret the plot:

- **Principal Component 1 Axis**: This axis represents the first principal component obtained from PCA. Each point on this axis corresponds to the projection of a data point onto the first principal component. The first principal component captures the direction of maximum variance in the original data.

- **Projected Data Points (Red)**: Each red point in the plot represents a data point after projecting the original data onto the first principal component. These points are the reduced-dimensional representation of the original data in 1D space. The position of each point along the Principal Component 1 axis represents its projection onto this component.

- **Interpretation**: The scatter plot shows how the original high-dimensional data has been compressed into 1D space while preserving the most significant variation. The distribution of the red points along the Principal Component 1 axis gives insight into the structure and variability of the dataset in this reduced-dimensional representation. Points that are closer together in the plot correspond to data points that are more similar in the original high-dimensional space, while points that are further apart represent greater dissimilarity.

In the context of PCA, the principal component axes themselves are not directly associated with any specific original features. Instead, each principal component axis is a linear combination of all the original features in the dataset.

The first principal component axis (PC1) represents the direction in the high-dimensional feature space along which the data exhibits the maximum variance. It is a new axis or direction that captures the most significant patterns or variability in the data.

Therefore, it's not accurate to say that a specific feature corresponds to the PC1 axis. Rather, PC1 is a weighted combination of all the original features, and each feature contributes differently to PC1 based on its covariance with other features in the dataset. The weights or coefficients associated with each original feature in the PC1 axis indicate their relative importance in defining the direction of maximum variance in the data.

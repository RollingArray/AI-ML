# Optimization Using Gradient Descent

In mathematical optimization, the goal is to find the minimum (or maximum) of a function. Gradient descent is an iterative optimization algorithm commonly used for this purpose.

#### 2.1 Step-size (7.1.1):

**Mathematical Notation:**
- Consider a function $ f(x) $ that you want to minimize.
- The gradient of $ f(x) $ with respect to $ x $ is denoted as $ \nabla f(x) $, representing the vector of partial derivatives.

**Gradient Descent Update Rule:**
- The basic gradient descent update rule is: $ x_{k+1} = x_k - \alpha \nabla f(x_k) $
  - $ x_{k+1} $: Updated value of the variable at iteration $ k+1 $.
  - $ x_k $: Current value of the variable at iteration $ k $.
  - $ \alpha $: Step size or learning rate, a positive constant determining the size of the steps.

**Interpretation:**
- The term $ \nabla f(x_k) $ points in the direction of the steepest increase of the function at $ x_k $.
- Multiplying it by $ \alpha $ scales the step size. A smaller $ \alpha $ means smaller steps, potentially preventing overshooting, while a larger $ \alpha $ may speed up convergence but risks overshooting the minimum.

#### 2.2 Gradient Descent With Momentum (7.1.2):

**Mathematical Notation:**
- Introducing a momentum term, typically denoted as $ \beta $, with $ 0 < \beta < 1 $.

**Modified Update Rule:**
- $ v_{k+1} = \beta v_k + (1 - \beta) \nabla f(x_k) $
- $ x_{k+1} = x_k - \alpha v_{k+1} $
  - $ v_{k+1} $: Momentum term.
  - The update now considers both the current gradient and the momentum from the previous iteration.

**Interpretation:**
- Momentum helps smooth out oscillations in the optimization process, as it takes into account the direction of the previous steps.
- The momentum term acts as a kind of memory, allowing the optimization algorithm to avoid getting stuck in local minima.

#### 2.3 Stochastic Gradient Descent (7.1.3):

**Mathematical Notation:**
- Consider a dataset $ D $ with $ N $ data points.

**Stochastic Gradient Descent Update Rule:**
- In each iteration, randomly select a subset of data points $ D' $ (mini-batch) and compute $ \nabla f(x_k, D') $.

**Modified Update Rule:**
- $ x_{k+1} = x_k - \alpha \nabla f(x_k, D') $
  - $ \nabla f(x_k, D') $: Gradient computed on the mini-batch.

**Interpretation:**
- Stochastic Gradient Descent (SGD) is faster than regular gradient descent, especially for large datasets, as it uses only a subset of data in each iteration.
- The randomness introduced by selecting mini-batches helps escape local minima and explores the parameter space more efficiently.

Understanding these mathematical formulations lays the foundation for implementing and customizing gradient descent algorithms for specific optimization problems. Practice with simple functions and gradually move to more complex ones to reinforce your understanding.

# Step Size

Certainly! $'s use a simple quadratic function to illustrate the concept of step size in gradient descent.

Consider the quadratic function:
$ f(x) = x^2 $

We want to minimize this function using gradient descent. The derivative of $ f(x) $ with respect to $ x $ is $ \frac{df}{dx} = 2x $. Now, $'s apply the gradient descent update rule:

$ x_{k+1} = x_k - \alpha \frac{df}{dx} $

In this case, $ \frac{df}{dx} = 2x $. Now, $'s choose an initial value $ x_0 = 5 $ and a small step size $ \alpha = 0.1 $ to see how the algorithm progresses in minimizing the function.

1. **Iteration 1:**
   $ x_1 = x_0 - 0.1 \cdot \frac{df}{dx}(x_0) = 5 - 0.1 \cdot 2 \cdot 5 = 4 $

2. **Iteration 2:**
   $ x_2 = x_1 - 0.1 \cdot \frac{df}{dx}(x_1) = 4 - 0.1 \cdot 2 \cdot 4 = 3.2 $

3. **Iteration 3:**
   $ x_3 = x_2 - 0.1 \cdot \frac{df}{dx}(x_2) = 3.2 - 0.1 \cdot 2 \cdot 3.2 = 2.56 $

And so on. The algorithm will continue to update the value of $ x $ in the negative direction of the gradient until it converges to the minimum.

**Observations:**
- A smaller step size ($ \alpha $) would result in smaller updates, potentially increasing the number of iterations needed for convergence.
- A larger step size may lead to faster convergence but risks overshooting the minimum and oscillations.

Experiment with different step sizes to see how they affect the convergence of the algorithm. It's a balance between choosing a step size that ensures convergence and avoiding overshooting.

$'s visualize the gradient descent process for the quadratic function $ f(x) = x^2$ using a small step size ($ \alpha = 0.1$).

````python
import numpy as np
import matplotlib.pyplot as plt

# Quadratic function
def f(x):
    return x**2

# Derivative of the quadratic function
def df_dx(x):
    return 2*x

# Gradient descent function
def gradient_descent(x, alpha, num_iterations):
    trajectory = [x]
    for _ in range(num_iterations):
        x = x - alpha * df_dx(x)
        trajectory.append(x)
    return np.array(trajectory)

# Initial values
x0 = 5
alpha = 0.1
num_iterations = 10

# Run gradient descent
trajectory = gradient_descent(x0, alpha, num_iterations)

# Plot the quadratic function
x_values = np.linspace(-6, 6, 100)
y_values = f(x_values)

# Plot the gradient descent trajectory
plt.plot(x_values, y_values, label='Quadratic Function')
plt.scatter(trajectory, f(trajectory), color='red', label='Gradient Descent Trajectory')
plt.title('Gradient Descent for Quadratic Function')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.show()
````

# Gradient Descent with Momentum:

### Gradient Descent Recap:

Gradient descent is an iterative optimization algorithm used to minimize a function. In the context of machine learning, this function often represents a measure of error or loss.

Mathematically, the basic update rule for gradient descent is:

$ x_{k+1} = x_k - \alpha \nabla f(x_k) $

Here:
- $ x_k$ is the current value of the parameter or variable.
- $ \alpha$ is the learning rate, a small positive constant.
- $ \nabla f(x_k)$ is the gradient of the function with respect to $ x$ at the current point.

### Gradient Descent with Momentum:

Gradient Descent with Momentum introduces a "momentum" term to the update rule. This momentum helps the optimization process by considering the direction of previous steps.

#### Update Rule with Momentum:

$ v_{k+1} = \beta v_k + (1 - \beta) \nabla f(x_k) $

$ x_{k+1} = x_k - \alpha v_{k+1} $

Here:
- $ v_k$ is the momentum term at iteration $ k$.
- $ \beta$ is a constant between 0 and 1, representing the momentum factor.
- The term $ (1 - \beta) \nabla f(x_k)$ represents the current gradient scaled by the momentum factor.
- The momentum term is updated by adding the scaled gradient to the previous momentum.

### Intuition:

1. **Memory Effect:**
   - Momentum introduces a "memory" to the optimization process. It remembers the direction of recent steps.

2. **Smoothing Oscillations:**
   - The momentum term helps smooth out oscillations in the optimization path, especially when the gradient changes direction frequently.

3. **Speeding up Convergence:**
   - Momentum can help accelerate convergence, especially through flat or shallow regions, by allowing the optimizer to "roll" along the surface.

### Mathematical Example:

$'s use a simple mathematical example to illustrate Gradient Descent with Momentum. Consider the function:

$ f(x) = x^2 $

This is a basic quadratic function, and we'll apply Gradient Descent with Momentum to minimize it.

1. **Objective Function:**
   $ f(x) = x^2 $

2. **Gradient of the Function:**
   $ \frac{df}{dx} = 2x $

3. **Gradient Descent with Momentum Update Rule:**
   $ v_{k+1} = \beta v_k + (1 - \beta) \frac{df}{dx}(x_k) $
   $ x_{k+1} = x_k - \alpha v_{k+1} $

$'s start with an initial value $ x_0 = 4 $, a learning rate $ \alpha = 0.1 $, and a momentum factor $ \beta = 0.9 $.

### Iterations:

1. **Iteration 0:**
   - $ x_0 = 4 $
   - $ v_0 = 0 $

2. **Iteration 1:**
   - $ \frac{df}{dx}(x_0) = 2 \times 4 = 8 $
   - $ v_1 = 0.9 \times 0 + (1 - 0.9) \times 8 = 0.8 $
   - $ x_1 = x_0 - 0.1 \times 0.8 = 4 - 0.08 = 3.92 $

3. **Iteration 2:**
   - $ \frac{df}{dx}(x_1) = 2 \times 3.92 = 7.84 $
   - $ v_2 = 0.9 \times 0.8 + (1 - 0.9) \times 7.84 = 1.568 $
   - $ x_2 = x_1 - 0.1 \times 1.568 = 3.92 - 0.1568 = 3.7632 $

4. **Iteration 3:**
   - Repeat the process...

### Visual Metaphor:

Imagine a ball rolling down a hill (representing the optimization process). Momentum helps the ball gather speed and navigate through flat areas, avoiding getting stuck.

This metaphorical understanding can help you grasp the essence of Gradient Descent with Momentum as a tool for efficient optimization.

````python
import numpy as np
import matplotlib.pyplot as plt

# Quadratic function
def f(x):
    return x**3

# Gradient of the quadratic function
def df_dx(x):
    return 3*x**2

# Gradient Descent with Momentum function
def gradient_descent_momentum(x, alpha, beta, num_iterations):
    trajectory = [x]
    v = 0  # Initial momentum

    for _ in range(num_iterations):
        gradient = df_dx(x)
        v = beta * v + (1 - beta) * gradient
        x = x - alpha * v
        trajectory.append(x)

    return np.array(trajectory)

# Initial values
x0 = 4
alpha = 0.1
beta = 0.9
num_iterations = 10

# Run Gradient Descent with Momentum
trajectory = gradient_descent_momentum(x0, alpha, beta, num_iterations)

# Plot the quadratic function
x_values = np.linspace(-6, 6, 100)
y_values = f(x_values)

# Plot the trajectory
plt.plot(x_values, y_values, label='Quadratic Function')
plt.scatter(trajectory, f(trajectory), color='red', label='Gradient Descent with Momentum')
plt.title('Gradient Descent with Momentum for Quadratic Function')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.show()
````

# Negative Gradient

The gradient of a function at a specific point represents the rate and direction of the steepest ascent of the function at that point. When the gradient is negative, it indicates the direction of the steepest descent or decrease of the function at that point.

Mathematically, for a function $f(x)$, the gradient (or derivative) at a point $x$ is denoted as $\frac{df}{dx}$. If $\frac{df}{dx}$ is negative at a particular point, it means that as you move along the x-axis in the positive direction from that point, the function is decreasing.

In the context of optimization, especially in algorithms like gradient descent, the negative gradient is essential. The negative gradient points in the direction of the steepest decrease of the function, and moving in this direction helps approach a local or global minimum of the function.

For example, in the context of the update rule in gradient descent:

$ x_{k+1} = x_k - \alpha \frac{df}{dx}(x_k) $

Here, $\frac{df}{dx}(x_k)$ is the gradient. If it's negative, subtracting it from $x_k$ in the update rule ensures that $x_{k+1}$ moves towards the direction of decreasing $f(x)$, aiming to find the minimum of the function.

In summary, a negative gradient is a key indicator in optimization, signifying the direction in which the function decreases most rapidly at a specific point.

Certainly! $'s use a simple mathematical example and then demonstrate the concept through a Python code snippet.

### Mathematical Example:

Consider the function $f(x) = x^2 - 4x + 4$. We'll find the points where the gradient is negative.

1. **Objective Function:**
   $ f(x) = x^2 - 4x + 4 $

2. **Gradient of the Function:**
   $ \frac{df}{dx} = 2x - 4 $

3. **Find where the Gradient is Negative:**
   Set $2x - 4 < 0$.
   $ 2x - 4 < 0 \implies x < 2 $

   So, the gradient is negative when $x < 2$.

### Python Code:

Now, $'s write a Python code snippet to visualize this function and mark the points where the gradient is negative.

# Stochastic Gradient Descent (SGD)
Absolutely! $'s dive into Stochastic Gradient Descent (SGD) with a simple mathematical example and a Python code snippet.

## Mathematical Example:

Consider the function $f(x) = x^2 - 4x + 4$. We want to use Stochastic Gradient Descent to minimize this function.

1. **Objective Function:**
   $ f(x) = x^2 - 4x + 4 $

2. **Gradient of the Function:**
   $ \frac{df}{dx} = 2x - 4 $

### Stochastic Gradient Descent (SGD) Algorithm:

The update rule for SGD is given by:
$ x_{k+1} = x_k - \alpha \nabla f(x_k, D') $

Here:
- $ D' $ is a mini-batch of the data.
- $ \alpha $ is the learning rate.
- $ \nabla f(x_k, D') $ is the gradient of the function with respect to $ x $ computed on the mini-batch.

### Iterations:

$'s perform a few iterations of SGD for this example:

1. **Initialization:**
   - $ x_0 = 2 $ (initial value)
   - $ \alpha = 0.1 $ (learning rate)
   - $ \text{batch size} = 1 $ (for simplicity, using a mini-batch of size 1)

2. **Iteration 1:**
   - Randomly choose a data point $ x_1 $ ($'s say $ x_1 = 4 $)
   - Compute $ \nabla f(x_0, D') = 2 \times 4 - 4 = 4 $
   - Update $ x_0 $: $ x_1 = x_0 - 0.1 \times 4 = 1.6 $

3. **Iteration 2:**
   - Randomly choose another data point $ x_2 $ ($'s say $ x_2 = -1 $)
   - Compute $ \nabla f(x_1, D') = 2 \times (-1) - 4 = -6 $
   - Update $ x_1 $: $ x_2 = x_1 - 0.1 \times (-6) = 1.6 + 0.6 = 2.2 $

Continue this process for several iterations.

### Interpretation:

SGD updates the parameter $ x $ using the gradient computed on a small random subset of the data at each iteration. This randomness helps escape local minima and explore the parameter space more efficiently.

Feel free to perform more iterations with different randomly chosen data points to observe how $ x $ converges towards the minimum of the function.

## Python Code:

Now, $'s implement SGD for the given function in Python.

This Python code generates a plot of the function $f(x) = x^2 - 4x + 4$ and shows the trajectory of SGD as it minimizes the function using random mini-batches.

Feel free to experiment with different initial values, learning rates ($ \alpha $), numbers of iterations, and batch sizes to observe how the SGD trajectory changes.

````python
import numpy as np
import matplotlib.pyplot as plt

# Define the function
def f(x):
    return x**2 - 4*x + 4

# Gradient of the function
def df_dx(x):
    return 2*x - 4

# Stochastic Gradient Descent (SGD) function
def stochastic_gradient_descent(x, alpha, num_iterations, batch_size):
    trajectory = [x]

    for _ in range(num_iterations):
        # Generate a random mini-batch
        random_batch = np.random.choice(np.arange(-2, 6), size=batch_size, replace=False)
        
        # Calculate the gradient on the mini-batch
        gradient = np.mean([df_dx(xi) for xi in random_batch])
        
        # Update x using SGD
        x = x - alpha * gradient
        
        trajectory.append(x)

    return np.array(trajectory)

# Initial value
x0 = 2
alpha = 0.1
num_iterations = 50
batch_size = 5  # Size of the mini-batch

# Run SGD
trajectory_sgd = stochastic_gradient_descent(x0, alpha, num_iterations, batch_size)

# Plot the function and the SGD trajectory
x_values = np.linspace(-2, 6, 100)
y_values = f(x_values)

plt.plot(x_values, y_values, label='Function: $x^2 - 4x + 4$')
plt.scatter(trajectory_sgd, f(np.array(trajectory_sgd)), color='red', label='SGD Trajectory')
plt.title('Stochastic Gradient Descent (SGD) for a Simple Function')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.show()
````

# Batch Gradient Descent
Batch Gradient Descent is a popular optimization algorithm used in machine learning to minimize a cost function by iteratively adjusting the model parameters. It involves computing the gradient of the cost function with respect to the parameters using the entire training dataset in each iteration. $'s break down the concept with a mathematical example and a simple Python program.

## Mathematical Example:

Certainly! $'s solve the batch gradient descent problem mathematically for a simple linear regression example.

### Linear Regression Cost Function:
Consider the cost function for linear regression:

$ J$theta) = \frac{1}{2m} \sum_{i=1}^{m} (h_\theta(x^{(i)}) - y^{(i)})^2 $

Here:
- $ \theta $ is the parameter vector.
- $ h_\theta(x^{(i)}) $ is the predicted value for input $ x^{(i)} $.
- $ y^{(i)} $ is the actual output for input $ x^{(i)} $.
- $ m $ is the number of training examples.

### Hypothesis Function:
The hypothesis function for linear regression is given by:

$ h_\theta(x) = \theta_0 + \theta_1 x $

### Update Rule for Batch Gradient Descent:
The update rule for batch gradient descent is given by:

$ \theta_j := \theta_j - \alpha \frac{1}{m} \sum_{i=1}^{m} (h_\theta(x^{(i)}) - y^{(i)})x_j^{(i)} $

Where:
- $ \alpha $ is the learning rate.
- $ j = 0, 1 $ for the two parameters $ \theta_0 $ and $ \theta_1 $.
- $ x_j^{(i)} $ is the $j$-th feature of the $i$-th training example.

### Applying the Update Rule:
$'s apply the update rule to find the optimal parameters $ \theta_0 $ and $ \theta_1 $ for a simple linear regression example.

### Step 1: Initialize Parameters:
$ \theta_0 = 0, \quad \theta_1 = 0 $

### Step 2: Update Rule:
$ \theta_0 := \theta_0 - \alpha \frac{1}{m} \sum_{i=1}^{m} (h_\theta(x^{(i)}) - y^{(i)}) $
$ \theta_1 := \theta_1 - \alpha \frac{1}{m} \sum_{i=1}^{m} (h_\theta(x^{(i)}) - y^{(i)})x^{(i)} $

### Step 3: Repeat:
Repeat Step 2 until convergence.

Please note that finding the analytical solution involves more complex calculus and linear algebra and is usually done using numerical methods or specialized optimization algorithms. The above steps provide an overview of the iterative process involved in batch gradient descent.

# Mini-Batch Gradient Descent
Mini-Batch Gradient Descent is a variation of the gradient descent algorithm that lies between the Batch Gradient Descent and Stochastic Gradient Descent approaches. In Mini-Batch Gradient Descent, instead of using the entire training dataset (as in Batch GD) or just a single data point (as in Stochastic GD), a small batch of data points is used in each iteration to update the model parameters. This approach combines some advantages of both the batch and stochastic methods.

### Mathematical Example:

Solving the Mini-Batch Gradient Descent (MBGD) problem mathematically involves finding the partial derivatives of the cost function with respect to the parameters and updating the parameters iteratively using the mini-batches of the training data. For simplicity, I'll provide a general outline of the mathematical steps involved in one iteration of Mini-Batch Gradient Descent for linear regression.

### Objective Function:
Consider the linear regression cost function:

$ J$theta) = \frac{1}{2b} \sum_{i=1}^{b} (h_\theta(x^{(i)}) - y^{(i)})^2 $

Where $ b $ is the batch size, $ \theta $ is the parameter vector, $ h_\theta(x^{(i)}) $ is the predicted value for input $ x^{(i)} $, $ y^{(i)} $ is the actual output for input $ x^{(i)} $.

### Hypothesis Function:
The hypothesis function for linear regression is given by:

$ h_\theta(x) = \theta_0 + \theta_1 x $

### Update Rule for Mini-Batch Gradient Descent:
The update rule for Mini-Batch Gradient Descent in one iteration is given by:

$ \theta_j := \theta_j - \alpha \frac{1}{b} \sum_{i=1}^{b} (h_\theta(x^{(i)}) - y^{(i)})x_j^{(i)} $

Where:
- $ \alpha $ is the learning rate.
- $ j = 0, 1 $ for the two parameters $ \theta_0 $ and $ \theta_1 $.
- $ b $ is the batch size.
- $ x_j^{(i)} $ is the $j$-th feature of the $i$-th training example in the mini-batch.

### Mathematical Steps:
1. Compute the partial derivatives of the cost function with respect to $ \theta_0 $ and $ \theta_1 $:

   $ \frac{\partial J}{\partial \theta_0} = \frac{1}{b} \sum_{i=1}^{b} (h_\theta(x^{(i)}) - y^{(i)}) $

   $ \frac{\partial J}{\partial \theta_1} = \frac{1}{b} \sum_{i=1}^{b} (h_\theta(x^{(i)}) - y^{(i)})x^{(i)} $

2. Update the parameters $ \theta_0 $ and $ \theta_1 $ using the update rule.

3. Repeat the process for multiple iterations until convergence.

Please note that finding the analytical solution for the entire optimization process involves complex calculus and linear algebra and is generally done using numerical methods or specialized optimization algorithms. The above steps provide an overview of the iterative process involved in Mini-Batch Gradient Descent.

### Python Program:

$'s implement Mini-Batch Gradient Descent for a simple linear regression problem:

This code demonstrates Mini-Batch Gradient Descent for a simple linear regression problem. The loss is tracked over iterations to visualize the convergence behavior. Adjust the learning rate, number of iterations, and batch size to observe different behaviors.

````python
import numpy as np
import matplotlib.pyplot as plt

# Generate synthetic data for a linear relationship
np.random.seed(100)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)

# Add a bias term to X
X_b = np.c_[np.ones((100, 1)), X]

# Mini-Batch Gradient Descent function with loss tracking
def mini_batch_gradient_descent(X, y, learning_rate=0.01, n_iterations=1000, batch_size=10):
    m = len(y)
    theta = np.random.randn(2, 1)  # Initialize parameters randomly
    loss_history = []

    for iteration in range(n_iterations):
        # Randomly select a mini-batch from the training data
        indices = np.random.randint(0, m, batch_size)
        X_batch = X[indices]
        y_batch = y[indices]

        gradients = 1/batch_size * X_batch.T.dot(X_batch.dot(theta) - y_batch)
        theta = theta - learning_rate * gradients

        # Calculate and store the loss
        loss = 1/(2*batch_size) * np.sum((X_batch.dot(theta) - y_batch)**2)
        loss_history.append(loss)

    return theta, loss_history

# Perform Mini-Batch Gradient Descent
theta_optimal, loss_history = mini_batch_gradient_descent(X_b, y)

# Plot the original data and the linear regression line
plt.scatter(X, y)
plt.plot(X, X_b.dot(theta_optimal), color='red', label='Linear Regression')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.show()

print("Optimal parameters (theta):", theta_optimal.flatten())

# Plot the loss over iterations
plt.plot(range(1, len(loss_history) + 1), loss_history, label='Loss')
plt.title('Loss over Iterations (Mini-Batch Gradient Descent)')
plt.xlabel('Iteration')
plt.ylabel('Loss')
plt.legend()
plt.show()
````

# Learning Rate Decay:

A constant learning rate can be problematic. If it's too low, the algorithm may take a long time to converge to an optimal solution. If it's too high, the algorithm may oscillate around the optimal solution without converging. Learning rate decay is introduced as a technique to dynamically adjust the learning rate over time.

The formula for updating the weight vector $w$ using a decaying learning rate $\alpha_t$ is given by:
$ w \leftarrow w - \alpha_t \nabla J $

Two common decay functions are mentioned:
1. Exponential Decay: $\alpha_t = \alpha_0 \exp(-k \cdot t)$
2. Inverse Decay: $\alpha_t = \alpha_0$

Here, $\alpha_0$ is the initial learning rate, $k$ controls the rate of decay, and $t$ represents the time measured in terms of the number of cycles over all training points.

# Bold Driver Algorithm:

The Bold Driver algorithm is proposed as an approach to dynamically adjust the learning rate based on the improvement or worsening of the objective function during the optimization process. The key steps are as follows:
- If the objective function improves, increase the learning rate by a factor (around 5%).
- If the objective function worsens, undo the step and retry with a reduced learning rate (by a factor of around 50%).

This process continues iteratively until convergence. It is noted that the Bold Driver algorithm may not work well in noisy settings, such as stochastic gradient descent, where the objective function is approximated using data samples. In such cases, it's suggested to test the objective function and adjust the learning rate after multiple steps (denoted as $m$) instead of a single step for more robust measurements.

In summary, learning rate decay and the Bold Driver algorithm are techniques designed to address the challenges of choosing an appropriate learning rate during the optimization process, ensuring a balance between convergence speed and stability.

## Mathematical Explanation 
Certainly! $'s consider a more complex function for illustration. We'll use the function $f(x) = 3x^3 - 6x^2 + 2x$ and apply the Bold Driver Algorithm to minimize it.

### Objective Function:
$ f(x) = 3x^3 - 6x^2 + 2x $

### Gradient of the Objective Function:
$ \nabla f(x) = 9x^2 - 12x + 2 $

### Bold Driver Algorithm:
$ \alpha_{t+1} = \beta_+ \cdot \alpha_t \quad \text{if the objective function improves} $
$ \alpha_{t+1} = \beta_- \cdot \alpha_t \quad \text{if the objective function worsens} $

````python
import numpy as np
import matplotlib.pyplot as plt

# Objective function (scaled down for numerical stability)
def complex_function(x):
    return 0.001 * (3*x**3 - 6*x**2 + 2*x)

# Gradient of the objective function
def gradient_complex_function(x):
    return 9*x**2 - 12*x + 2

# Bold Driver Algorithm
def bold_driver_algorithm_complex(starting_point, initial_learning_rate, beta_plus, beta_minus, max_iterations):
    current_position = starting_point
    learning_rate = initial_learning_rate
    trajectory = $current_position$

    for iteration in range(1, max_iterations + 1):
        current_value = complex_function(current_position)
        gradient = gradient_complex_function(current_position)
        
        new_position = current_position - learning_rate * gradient
        new_value = complex_function(new_position)

        if new_value < current_value:
            # If the objective function improves, increase the learning rate
            learning_rate *= beta_plus
        else:
            # If the objective function worsens, decrease the learning rate
            learning_rate *= beta_minus

        current_position = new_position
        trajectory.append(current_position)

        print(f"Iteration {iteration}: x_{iteration} = {current_position:.4f}, alpha_{iteration} = {learning_rate:.4f}")

    return trajectory

# Plot the complex function and trajectory with Bold Driver Algorithm
initial_learning_rate = 0.1
beta_plus = 1.05
beta_minus = 0.5
starting_point = 2
max_iterations = 10

# Perform Bold Driver Algorithm
trajectory = bold_driver_algorithm_complex(starting_point, initial_learning_rate, beta_plus, beta_minus, max_iterations)

# Plot the complex function
x_values = np.linspace(-2, 3, 100)
y_values = complex_function(x_values)

# Plot the trajectory
plt.plot(x_values, y_values, label='Complex Function: $0.001(x^4 + 3x^3 - 6x^2 + 2x)$')
plt.scatter(trajectory, $complex_function(x) for x in trajectory$, color='red', label='Trajectory with Bold Driver')
plt.title('Bold Driver Algorithm with Complex Function')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.show()
````

# Line Search
The "Line Search" method is a technique used in optimization algorithms to find the optimal step size along a given direction in order to minimize the objective function. In this method, the learning rate (step size) is chosen in each step to minimize the value of the objective function at the new parameter vector.

$ J$alpha) = 2\alpha^2 - 3\alpha + 5 $

### Step 1: Compute the Derivative:

$ J'$alpha) = \frac{dJ}{d\alpha} = 4\alpha - 3 $

### Step 2: Set the Derivative to Zero:

Setting $ J'$alpha) $ to zero and solving for $ \alpha $, we get:

$ 4\alpha - 3 = 0 $
$ 4\alpha = 3 $
$ \alpha = \frac{3}{4} $

So, $ \alpha = \frac{3}{4} $ is a critical point.

### Step 3: Check for Minimum:

To determine if $ \alpha = \frac{3}{4} $ corresponds to a minimum, we can use the second derivative test or evaluate the function at nearby points.

### Step 4: Evaluate the Function:

$'s evaluate the objective function at $ \alpha = \frac{3}{4} $:

$ J\left$frac{3}{4}\right) = 2\left$frac{3}{4}\right)^2 - 3\left$frac{3}{4}\right) + 5 = \frac{9}{8} - \frac{9}{4} + 5 = \frac{9}{8} - \frac{18}{8} + \frac{40}{8} = \frac{9 - 18 + 40}{8} = \frac{31}{8} $

So, $ J\left$frac{3}{4}\right) = \frac{31}{8} $.

Therefore, the minimum value of the objective function occurs at $ \alpha = \frac{3}{4} $, and the minimum value of the function is $ \frac{31}{8} $.

# Binary Search for Optimization:

1. **Initialization:**
   - Start by initializing the binary search interval for the parameter $\alpha$ to $[a, b] = [0, \alpha_{\text{max}}]$.

2. **Binary Search Iterations:**
   - In each iteration, evaluate the objective function at two closely spaced points near the midpoint $(a + b)/2$:
     - Evaluate $H\left$frac{a+b}{2}\right)$ and $H\left$frac{a+b}{2} + \varepsilon\right)$, where $\varepsilon$ is a small value (e.g., $10^{-6}$).
   
   - Determine whether the objective function is increasing or decreasing at $(a + b)/2$ by comparing the two evaluations:
     - If $H\left$frac{a+b}{2}\right) < H\left$frac{a+b}{2} + \varepsilon\right)$, the function is increasing at $(a + b)/2$.
     - If $H\left$frac{a+b}{2}\right) \geq H\left$frac{a+b}{2} + \varepsilon\right)$, the function is not increasing at $(a + b)/2$.

3. **Update the Interval:**
   - Narrow the interval based on the evaluation:
     - If the function is increasing at $(a + b)/2$, set the new interval to $[a, (a + b)/2 + \varepsilon]$.
     - If the function is not increasing at $(a + b)/2$, set the new interval to $[(a + b)/2, b]$.

4. **Repeat:**
   - Repeat the process until the interval reaches the required level of accuracy or until a termination criterion is met.

## Mathematical example:
$'s mathematically solve the complex quadratic function $f(x) = x^2 + 3x + 5$ using calculus. We want to find the critical points of the function, where the derivative is equal to zero.

### Objective Function:
$ f(x) = x^2 + 3x + 5 $

### Derivative:
$ f'(x) = 2x + 3 $

### Set Derivative to Zero:
$ 2x + 3 = 0 $

### Solve for $x$:
$ x = -\frac{3}{2} $

The critical point $x = -\frac{3}{2}$ is the only point where the derivative is equal to zero. To determine whether this point is a minimum, maximum, or saddle point, we can use the second derivative test.

### Second Derivative:
$ f''(x) = 2 $

Since the second derivative is positive, the critical point $x = -\frac{3}{2}$ corresponds to a local minimum.

Now, $'s implement this in Python and visualize the result:

This code plots the objective function, its derivative, and marks the critical point. As expected, the critical point corresponds to the minimum of the function.

````python
import numpy as np
import matplotlib.pyplot as plt

# Objective function (replace this with your actual objective function)
def objective_function(alpha):
    return (alpha - 2)**2

# Binary search for optimization
def binary_search_optimization(a, b, epsilon, alpha_max):
    while (b - a) > epsilon:
        mid = (a + b) / 2
        mid_value = objective_function(mid)
        mid_epsilon_value = objective_function(mid + epsilon)

        if mid_value < mid_epsilon_value:
            b = mid + epsilon
        else:
            a = mid

    optimal_alpha = (a + b) / 2
    return optimal_alpha

# Plot the objective function and the optimization process
alpha_max = 5
epsilon = 1e-6

# Perform binary search optimization
optimal_alpha = binary_search_optimization(0, alpha_max, epsilon, alpha_max)

# Plot the objective function
alpha_values = np.linspace(0, alpha_max, 100)
f_values = objective_function(alpha_values)

plt.plot(alpha_values, f_values, label='Objective Function: $$\alpha - 2)^2$')
plt.scatter(optimal_alpha, objective_function(optimal_alpha), color='red', label='Optimal $\\alpha$ after Binary Search')
plt.title('Binary Search Optimization')
plt.xlabel('$\\alpha$')
plt.ylabel('$f$\alpha)$')
plt.legend()
plt.grid(True)
plt.show()

print(f"Optimal $\\alpha$: {optimal_alpha}")
````

# Golden-Section Search:

Based on the given information, here's an explanation of the Golden-Section Search algorithm:

**Similarities to Binary Search:**

* Starts with an initial interval `[a, b]`. This is similar to binary search, where you start with the entire search space.

**Core Difference:**

* **Binary Search:** Divides the interval in half at each step.
* **Golden-Section Search:** Divides the interval using a specific ratio based on the **golden ratio** (approximately 1.618).

**Golden-Section Search Steps:**

1. **Divide the Interval:**
    * Calculate two points within the interval `[a, b]`:
        * `m1 = a + (b - a) * ( - 1)` ( represents the golden ratio)
        * `m2 = b - (b - a) * ( - 1)`
2. **Evaluate the Function:**
    * Evaluate the function `H()` at each point: `H(m1)`, `H(m2)`.
3. **Discard Large Interval:**
    * Compare the function values:
        * If `H(m1) < H(m2)`, discard the interval `[m2, b]`.
        * If `H(m1) > H(m2)`, discard the interval `[a, m1]`.
    * In some cases, an even larger interval can be discarded (shown in the text, but not crucial for understanding the basic concept).
4. **Update Interval:**
    * Reset the interval `[a, b]` based on the discarded portion.

**Key Point:**

* This process guarantees that at least one of the discarded intervals contains no minimum point.
* We are essentially **zooming in** on the region where the minimum likely resides.

**Additional Considerations:**

* **Initial Random Points:** 
    * If the interval after discarding sections has no evaluated points, the algorithm chooses two random points within the remaining interval.
* **Single Evaluated Point:** 
    * If the interval has only one evaluated point, another point is chosen within the larger half of the remaining interval.
* **Termination:**
    * The process repeats until the interval size becomes small enough (desired accuracy level).

**Advantages:**

* More efficient than repeatedly dividing the interval in half (like binary search).
* Fewer function evaluations needed to achieve the same level of accuracy.

**Disadvantages:**

* Slower than some other optimization methods.
* May not converge to the global minimum if the function is not unimodal (has multiple peaks and valleys).

**In essence, the Golden-Section Search leverages the properties of the golden ratio to strategically discard portions of the search space, efficiently converging towards the minimum point of a unimodal function.**

# Loss Function

* **Intuitive Explanation:** Imagine you're playing a game of darts. The Loss Function represents the distance between your dart (prediction) and the bull's eye (actual value). The closer your darts are to the bull's eye (lower loss), the better you're performing in the game (better model performance).

**Mathematical Example: Mean Squared Error (MSE):**

* **Analogy:** Similar to calculating the average squared distance of your darts from the bull's eye.
* **Formula:**

```
MSE = (1/n) *  (y_i - y_i)^2
```

- **Breakdown:**
    - `n`: Number of data points (darts thrown)
    - `y_i`: Actual value (bull's eye position)
    - `y_i`: Predicted value (your dart's landing position)
    - `(y_i - y_i)`: Difference between actual and predicted value (distance from bull's eye)
    - `(y_i - y_i)^2`: Squared difference (emphasizes larger errors)
    - ``: Summation over all data points (total distance from all darts)
    - `(1/n)`: Average squared difference (average distance from bull's eye)

**Python Code for Visualization (MSE):**

**Explanation of the Code:**

1. **Sample Data:** Creates data points (`x`) and assigns actual values (`y_true`) based on a linear relationship.
2. **Model Prediction:** Defines model parameters (`w` and `b`) and calculates predicted values (`y_pred`) using a simple linear model.
3. **MSE Calculation:** Calculates the Mean Squared Error (`mse`) using the formula.
4. **Visualization:**
    - Plots actual values (blue dots) and predicted values (red line).
    - Plots green dashed lines to represent the absolute error (difference) between each data point.
    - Adds labels, title, grid, and legend for clarity.
5. **Optional:** Identifies the data point with the minimum squared error (closest to the prediction line) and highlights it with a purple marker.

**Key Points:**

- This visualization depicts the absolute errors between actual and predicted values, along with the overall MSE loss.
- Lower absolute errors and a smaller MSE value indicate a better fit between the model and the data.

By understanding and visualizing loss functions, you gain valuable insights into how well your model performs and how much improvement is needed during the training process.


````python
import matplotlib.pyplot as plt
import numpy as np

# Sample data (representing actual target values)
x = np.linspace(0, 5, 100)  # Creates 100 data points between 0 and 5
y_true = x**4 + 5*x**3 + 12*x**2 + 2 * x + 3  # Actual values based on a linear relationship

# Model prediction (assuming a simple linear model)
w = 1  # Model parameter (slope)
b = 2  # Model parameter (y-intercept)
y_pred = w * x + b  # Predicted values based on the model

# Calculate MSE loss
mse = np.mean((y_true - y_pred)**2)

# Plot data and prediction
plt.scatter(x, y_true, label="Actual Values", color='blue')  # Plot actual values as blue dots
plt.plot(x, y_pred, label="Predicted Values", color='red')  # Plot predicted values as a red line

# Visualize loss as vertical lines (absolute errors)
for i in range(len(x)):
    plt.plot([x[i], x[i]], [y_true[i], y_pred[i]], color='green', linestyle='--', alpha=0.5)  # Green dashed lines with partial transparency

# Additional elements for clarity
plt.title("Loss Function Visualization (Mean Squared Error)")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.grid(True)  # Add grid lines for better visualization

# Highlight minimum MSE point (optional)
min_idx = np.argmin((y_true - y_pred)**2)  # Index of the data point with the lowest squared error
plt.scatter(x[min_idx], y_true[min_idx], marker='o', color='purple', label="Minimum Loss Point")  # Plot the point with minimum loss

plt.show()
````

# Overfitting

- It occurs when a model learns the training data too well, including its noise and peculiarities, to the extent that it performs poorly on new, unseen data.
- Imagine a student memorizing answers for a test without understanding the concepts. They might ace that specific test but struggle with different questions on a new exam.

- Consider a polynomial regression model with different degrees:


    - Using simpler models
    - Tuning model parameters (e.g., regularization)
    - Using more training data
    - Employing techniques like cross-validation

# Bias-Variance Tradeoff:

- A fundamental concept in machine learning that deals with the balance between two sources of error:

   - **Bias:** Represents the systematic underestimation or overestimation of the true relationship between features and target variables. 
     - Imagine a seesaw that consistently tilts to one side, preventing accurate balancing.

   - **Variance:** Captures the model's sensitivity to the specific training data. 
     - A model with high variance might perform well on the specific training data it was trained on but struggle to generalize to unseen data, similar to a seesaw that's overly sensitive to small movements.

## Mathematical Example:

While a specific mathematical formula isn't directly used to quantify the tradeoff, we can analyze how bias and variance affect the expected prediction error:

**Expected Prediction Error (EPE):**

```
EPE = Bias^2 + Variance
```

- Minimizing EPE is desirable for good model performance.
- **Tradeoff:** Reducing bias often leads to increased variance, and vice versa.

**Simple Analogy:**

Imagine throwing darts at a target.

- **High Bias (Underfitting):** Darts consistently land far from the bull's eye due to an inherent flaw in the throwing technique (bias).
- **High Variance (Overfitting):** Darts land scattered around the board due to inconsistency in throws, even if some hit close to the bull's eye (variance).
- **Ideal Scenario:** Darts consistently cluster around the bull's eye (low bias and variance).

# Normalization techniques

Normalization techniques are used to scale and transform the features of a dataset to ensure that they have similar ranges or distributions. This is particularly important for machine learning algorithms that are sensitive to the scale of the input features, such as gradient descent-based optimization algorithms.

Here are two commonly used normalization techniques:

1. Min-Max Scaling:
   - **Mathematical Explanation**: Min-max scaling, also known as min-max normalization, rescales the features to a fixed range, typically [0, 1]. It is done by subtracting the minimum value of the feature and then dividing by the range of the feature.
   - For a feature $ x $, the min-max scaled value $ x' $ is calculated as:
     $ x' = \frac{x - \text{min}(x)}{\text{max}(x) - \text{min}(x)} $
   - This transformation ensures that all feature values lie within the range [0, 1].
   
2. Standardization (Z-score normalization):
   - **Mathematical Explanation**: Standardization transforms the features so that they have a mean of 0 and a standard deviation of 1. It is done by subtracting the mean of the feature and then dividing by the standard deviation of the feature.
   - For a feature $ x $, the standardized value $ x' $ is calculated as:
     $ x' = \frac{x - \text{mean}(x)}{\text{std}(x)} $
   - This transformation centers the feature distribution around 0 and scales it to have unit variance.
  
3. **Mean Centering**:
   - **Mathematical Explanation**: Mean centering subtracts the mean of each feature from the data, resulting in a mean of 0 for each feature.
   - For a feature $ x $, the mean-centered value $ x' $ is calculated as:
     $ x' = x - \text{mean}(x) $
   - This transformation centers the distribution of each feature around 0.
   
4. **Standard Deviation Normalization**:
   - **Mathematical Explanation**: Standard deviation normalization divides each feature by its standard deviation, resulting in a standard deviation of 1 for each feature.
   - For a feature $ x $, the standard deviation normalized value $ x' $ is calculated as:
     $ x' = \frac{x}{\text{std}(x)} $
   - This transformation scales each feature to have unit variance.
   
5. **Batch Normalization Layer**:
   - Batch normalization is a layer typically inserted into deep neural networks between layers. It normalizes the activations of each layer across the mini-batch during training.
   - **Mathematical Explanation**: Batch normalization normalizes the inputs of each layer to have zero mean and unit variance across the mini-batch.
   - It consists of two learnable parameters, gamma ($ \gamma $) and beta ($ \beta $), which scale and shift the normalized inputs, allowing the model to learn the optimal normalization for each layer.
   - Batch normalization helps to stabilize and accelerate the training process by reducing internal covariate shift and enabling higher learning rates.

# Local optima and flat regions
Local optima and flat regions are two phenomena often encountered in optimization problems, particularly in non-convex optimization landscapes. Let's define each:

1. **Local Optima**:
   - In optimization, a local optimum is a point where the objective function has the lowest value within a local neighborhood but may not be the global minimum over the entire search space.
   - Mathematically, a point $ x^* $ is a local optimum if there exists a neighborhood $ N(x^*) $ such that for all $ x $ in $ N(x^*) $, $ f(x^*) \leq f(x) $, where $ f(x) $ is the objective function.
   - Local optima can be problematic because optimization algorithms, especially gradient-based methods, may converge to these points and fail to find the globally optimal solution.
   
2. **Flat Regions**:
   - A flat region, also known as a plateau or saddle, is an area in the optimization landscape where the objective function has nearly constant values over a range of inputs.
   - In a flat region, the gradient of the objective function becomes very small or even zero, making it difficult for optimization algorithms to make progress.
   - Flat regions can hinder optimization convergence because the algorithm may struggle to escape these regions and continue towards the optimal solution.
   
Both local optima and flat regions pose challenges for optimization algorithms, especially in high-dimensional and non-convex optimization problems. Techniques such as multi-start optimization, random restarts, and adaptive learning rates are often employed to mitigate the effects of local optima, while more sophisticated methods like second-order optimization and stochastic gradient descent variants can help in navigating flat regions.

# Differential Survature

1. **Vanishing and Exploding Gradient Problem**:
   - In complex optimization landscapes, such as those encountered in training neural networks, the gradients of the objective function (loss) may vary significantly in magnitude across different dimensions or parameters.
   - This variation can lead to problems like vanishing gradients (where gradients become very small) or exploding gradients (where gradients become very large), especially in deep neural networks with many layers.
   - Even in simpler cases, like convex and quadratic objective functions, minor manifestations of this problem may occur.

2. **Understanding the Concept**:
   - To understand this concept better, we start by examining simple cases, such as convex, quadratic objective functions, which provide valuable insights into the source of the problem and potential solutions.
   - Imagine a bowl-shaped surface representing the loss function, where the objective is to find the lowest point (global minimum). The curvature of this surface, particularly in different directions, affects the behavior of gradient descent.
  
3. **Illustrative Example**:
   **Two valleys (loss functions) are presented:**

4. **Circular valley (L = x^2 + y^2):**
   - This valley is like a perfect bowl.
   - All directions (changes in x and y) have similar impact on reaching the bottom (minimum).
   - Gradient descent (a method for finding the minimum) can reach the bottom quickly.

5. **Elliptical valley (L = x^2 + 4y^2):**
   - This valley is like an oval-shaped bowl.
   - Changes in the y direction have a larger impact on the landscape compared to changes in x.
   - Imagine a steeper slope along the y-axis and a gentler slope along the x-axis.
   - Gradient descent might:
     - Take large steps along the steeper y-direction.
     - Need to constantly "correct" its path due to the uneven slopes.
     - Make slow progress overall.
- - The difference in second-order derivatives ($\frac{\partial^2 L}{\partial x^2}$ and $\frac{\partial^2 L}{\partial y^2}$) indicates varying curvature, influencing how quickly gradients change.

1. **Implications for Gradient Descent**:
   - High curvature (high second-order derivatives) implies rapid changes in gradients, affecting the speed and direction of gradient descent.
   - Directions with more consistent gradients (less rapid changes) are preferred for making larger gradient-descent steps.

# Feature Normalization

3. **Linear Regression Model**: We want to predict the happiness index (target variable) based on the guns and butter features. In a linear regression model, we use coefficients (w1 for guns and w2 for butter) to predict the happiness index using the equation:  
   y = w1 * x1 + w2 * x2

4. **Objective Function**: We formulate the least-squares objective function, which measures the error between the actual and predicted happiness indices. The objective function is a function of the coefficients w1 and w2. It's represented as:  
   J = (0.1w1 + 25w2 - 7)^2 + (0.8w1 + 10w2 - 1)^2 + (0.4w1 + 10w2 - 4)^2

# Momentum-based Learning
This excerpt discusses momentum-based learning in the context of adjusting first-order derivatives for descent in optimization algorithms. Let's break down the key points:

1. **Introduction to Momentum-Based Learning**:
   - Momentum-based methods aim to address challenges such as local optima, flat regions, and zigzagging in the loss surface during gradient descent optimization.
   - These methods emphasize medium- to long-term directions of consistent movement by incorporating an aggregated measure of feedback from previous steps, akin to using momentum to overcome minor obstacles while rolling down a slope.

2. **Mathematical Formulation**:
   - In standard gradient descent updates, the parameter vector $ W $ is updated using the negative gradient multiplied by the learning rate $ \alpha $:
     $ V \leftarrow -\alpha \frac{\partial J}{\partial W}; \quad W \leftarrow W + V $
   - Momentum-based updates modify this process by incorporating a fraction $ \beta $ of the velocity from the previous step, in addition to the current gradient:
     $ V \leftarrow \beta V - \alpha \frac{\partial J}{\partial W}; \quad W \leftarrow W + V $
   - Here, $ \beta $ is the momentum parameter, typically chosen to be between 0 and 1, controlling the influence of the accumulated velocity from previous steps.

3. **Effect of Momentum**:
   - Momentum helps gradient descent navigate through flat regions and escape local optima by retaining speed and accumulating momentum in the correct direction.
   - The analogy of a marble rolling down a bowl illustrates how momentum-based methods overshoot slightly in the direction of accumulated velocity, leading to faster convergence.
   - Properly chosen momentum values help smooth out zigzagging caused by contradictory steps, allowing larger steps in the correct direction without causing oscillations.

4. **Acceleration of Learning**:
   - By favoring consistent directions over multiple steps and muting useless sideways oscillations, momentum-based descent accelerates learning and facilitates convergence to the optimal solution.
   - The concept is further illustrated by visualizing the movement of a marble down a gently sloping valley, where the effects of bouncing along the sides of the valley are gradually muted as momentum accumulates.

In summary, momentum-based learning modifies the gradient descent process by incorporating momentum from previous steps, enabling faster convergence and more efficient optimization in the presence of complex loss surfaces.

````python
import numpy as np
import matplotlib.pyplot as plt

# Define the loss function (bowl-shaped)
def loss_function(x):
    return x**2

# Define gradient descent function without momentum
def gradient_descent_without_momentum(learning_rate, num_iterations):
    trajectory = []
    x = 2.0  # Starting point
    for _ in range(num_iterations):
        gradient = 2 * x  # Gradient of the loss function
        x -= learning_rate * gradient
        trajectory.append(x)
    return trajectory

# Define gradient descent function with momentum
def gradient_descent_with_momentum(learning_rate, momentum, num_iterations):
    trajectory = []
    x = 2.0  # Starting point
    velocity = 0  # Initial velocity
    for _ in range(num_iterations):
        gradient = 2 * x  # Gradient of the loss function
        velocity = momentum * velocity - learning_rate * gradient
        x += velocity
        trajectory.append(x)
    return trajectory

# Generate data for plotting
x_vals = np.linspace(-3, 3, 100)
y_vals = loss_function(x_vals)
trajectory_no_momentum = gradient_descent_without_momentum(0.1, 10)
trajectory_with_momentum = gradient_descent_with_momentum(0.1, 0.9, 10)

# Plot gradient descent without momentum
plt.figure(figsize=(10, 6))
plt.plot(x_vals, y_vals, label='Loss Surface')
plt.plot(trajectory_no_momentum, loss_function(np.array(trajectory_no_momentum)), marker='o', label='Without Momentum')
plt.xlabel('Parameter Value')
plt.ylabel('Loss')
plt.title('Gradient Descent without Momentum')
plt.legend()
plt.grid(True)
plt.show()

# Plot gradient descent with momentum
plt.figure(figsize=(10, 6))
plt.plot(x_vals, y_vals, label='Loss Surface')
plt.plot(trajectory_with_momentum, loss_function(np.array(trajectory_with_momentum)), marker='o', label='With Momentum')
plt.xlabel('Parameter Value')
plt.ylabel('Loss')
plt.title('Gradient Descent with Momentum')
plt.legend()
plt.grid(True)
plt.show()
````

# AdaGrad

AdaGrad is an optimization algorithm commonly used in machine learning for training models. It is particularly useful when dealing with sparse data and can adaptively adjust the learning rate for each parameter based on the historical gradients.

Here's how AdaGrad works:

1. **Aggregated Squared Magnitude of Gradients:** In AdaGrad, for each parameter $ w_i $, the algorithm keeps track of the aggregated squared magnitude of the partial derivative with respect to that parameter over the course of the training process. Let's denote this aggregated value as $ A_i $.

2. **Update Rule:** In each iteration of the algorithm, AdaGrad updates the aggregated squared magnitude $ A_i $ for each parameter by adding the square of the current partial derivative $ $partial J / \partial w_i)^2 $:
   $ A_i \leftarrow A_i + $partial J / \partial w_i)^2 $

3. **Adjusting Learning Rate:** AdaGrad adjusts the learning rate for each parameter by scaling it inversely with the square root of the aggregated squared magnitude $ A_i $. This scaling helps in normalizing the gradients and encourages faster relative movements along gently sloping directions with consistent signs of the gradient.

4. **Parameter Update:** Finally, AdaGrad updates each parameter $ w_i $ using the adjusted learning rate:
   $ w_i \leftarrow w_i - \frac{\eta}{\sqrt{A_i + \epsilon}} \cdot \frac{\partial J}{\partial w_i} $
   Here, $ \eta $ is the learning rate, and $ \epsilon $ is a small positive value (like $ 10^{-8} $) added to the denominator to prevent division by zero.

By adapting the learning rates based on the historical gradients, AdaGrad can effectively handle sparse data and converge faster than traditional gradient descent algorithms. However, AdaGrad has some limitations, such as the tendency for the learning rates to become too small over time, which can lead to premature convergence. This issue is addressed by algorithms like RMSProp, which use exponential averaging to adjust the learning rates.

## Mathematical example
Let's derive the update rule for AdaGrad for a general optimization problem.

Given an objective function $ J(w) $, where $ w $ is the parameter vector, the update rule for AdaGrad is as follows:

$
w_i \leftarrow w_i - \frac{\eta}{\sqrt{A_i + \epsilon}} \cdot \frac{\partial J}{\partial w_i}
$

Where:
- $ w_i $ is the $ i $-th parameter,
- $ \eta $ is the learning rate,
- $ A_i $ is the aggregated squared magnitude of the partial derivative $ \frac{\partial J}{\partial w_i} $,
- $ \epsilon $ is a small positive constant added for numerical stability.

The aggregation of squared gradients $ A_i $ is updated at each iteration as follows:

$
A_i \leftarrow A_i + \left( \frac{\partial J}{\partial w_i} \right)^2
$

This update accumulates the squared gradients over time.

Now, let's consider a simple example of optimizing a quadratic function:

$
J(w) = w^2
$

The gradient of this function with respect to parameter $ w $ is:

$
\frac{\partial J}{\partial w} = 2w
$

Using AdaGrad, we can derive the update rule for this specific case:

$
w \leftarrow w - \frac{\eta}{\sqrt{A + \epsilon}} \cdot (2w)
$

where $ A $ is the aggregated squared magnitude of the gradient.

This equation represents the update rule for AdaGrad applied to the quadratic function optimization problem.

Let's consider a slightly more complex function:

$ J(w) = w^3 - 2w^2 + 3w + 1 $

The derivative of this function with respect to $ w $ is:

$ \frac{dJ}{dw} = 3w^2 - 4w + 3 $

Now, let's apply AdaGrad to update the parameter $ w $. We'll assume an initial value of $ w = 2 $, a learning rate $ \eta = 0.05 $, and $ \epsilon = 10^{-6} $. We'll also initialize $ A $ to zero.

1. **Initialization**:
   - $ w = 2 $
   - $ \eta = 0.05 $
   - $ \epsilon = 10^{-6} $
   - $ A = 0 $

2. **Calculate Gradient**:
   - $ \frac{dJ}{dw} = 3 \cdot 2^2 - 4 \cdot 2 + 3 = 12 - 8 + 3 = 7 $

3. **Update $ A $**:
   - $ A = A + (7)^2 = 49 $

4. **Update $ w $**:
   - $ w = w - \frac{\eta}{\sqrt{A + \epsilon}} \cdot \frac{dJ}{dw} $
   - $ w = 2 - \frac{0.05}{\sqrt{49 + 10^{-6}}} \cdot 7 $
   - $ w \approx 2 - \frac{0.05}{\sqrt{49}} \cdot 7 $
   - $ w \approx 2 - \frac{0.05}{7} \cdot 7 $
   - $ w \approx 2 - 0.05 $
   - $ w \approx 1.95 $

5. **Repeat**:
   - Repeat steps 2-4 for the desired number of iterations.

This demonstrates the manual steps of applying AdaGrad to update the parameter $ w $ for a slightly more complex cubic function.

## Python code

To visualize how AdaGrad works, let's create a simple example where we optimize a quadratic function using AdaGrad. We'll plot the optimization path to observe how the algorithm adapts the learning rates for each parameter.

In this visualization:
- The x-axis represents the parameter value.
- The y-axis represents the function value.
- Each point on the plot corresponds to the parameter value and the function value after each iteration of AdaGrad optimization.
- The optimization path shows how AdaGrad adapts the learning rates over iterations to minimize the quadratic function.

You can observe how AdaGrad efficiently adjusts the learning rates based on the historical gradients, allowing it to converge towards the optimal solution of the quadratic function.

````python
import numpy as np
import matplotlib.pyplot as plt

# Define the quadratic function
def quadratic_function(x):
    return x**2

# Initialize parameters
learning_rate = 0.1
epsilon = 1e-8
num_iterations = 100

# Initialize parameters for AdaGrad
A = 0

# Lists to store parameter values and function values
parameters = []
function_values = []

# Perform AdaGrad optimization
for i in range(num_iterations):
    # Compute gradient
    gradient = 2 * i
    
    # Update aggregated squared magnitude
    A += gradient**2
    
    # Adjusted learning rate using AdaGrad
    adjusted_learning_rate = learning_rate / np.sqrt(A + epsilon)
    
    # Update parameter using adjusted learning rate
    parameters.append(-adjusted_learning_rate * gradient)  # Minimize the function, so negate the gradient
    function_values.append(quadratic_function(parameters[-1]))

# Plot the optimization path
plt.figure(figsize=(10, 6))
plt.plot(parameters, function_values, marker='o', linestyle='-')
plt.xlabel('Parameter Value')
plt.ylabel('Function Value')
plt.title('AdaGrad Optimization Path for Quadratic Function')
plt.grid(True)
plt.show()
````

# RMSProp
RMSProp, which stands for Root Mean Square Propagation, is another optimization algorithm commonly used in training neural networks. It addresses some of the limitations of AdaGrad, particularly the issue of premature slowing down of learning due to the ever-increasing scaling factor $ A_i $.

Here's a breakdown of how RMSProp works:

1. **Exponential Averaging**: Instead of simply adding the squared gradients to estimate $ A_i $ like in AdaGrad, RMSProp uses exponential averaging. It keeps a running estimate of the second moment of the gradients.

2. **Decay Factor $ \rho $**: RMSProp introduces a decay factor $ \rho $ (rho) between 0 and 1. This factor controls how much weight to give to recent gradients versus older ones. Typically, $ \rho $ is set close to 1, meaning recent gradients have more influence.

3. **Updating $ A_i $**: The update equation for $ A_i $ in RMSProp is as follows:
   $ A_i \leftarrow \rho A_i + (1 - \rho) $partial w)^2 $
   Here, $ \partial w $ represents the gradient of the parameter $ w $ with respect to the loss function.

4. **Normalization**: The square root of the $ A_i $ values is used to normalize the gradient for each parameter. This ensures that parameters with large gradients are appropriately scaled.

5. **Learning Rate Update**: The learning rate $ \alpha $ is then divided by the square root of $ A_i $ to adjust the step size for each parameter update.

6. **Bias Correction**: Since $ A_i $ is initialized to 0 and updated iteratively, there's a bias in the early iterations. This bias disappears over time as the running estimate becomes more accurate. This bias correction ensures that early iterations are not overly influenced by the initial estimates of $ A_i $.

Overall, RMSProp provides a more balanced approach to adapting the learning rate based on the magnitude of gradients. It prevents the learning rate from becoming too small too quickly, as can happen in AdaGrad. Additionally, it addresses the issue of stale gradients by exponentially decaying the influence of older gradients over time.

## Mathematical example:
Sure, let's consider the following function:

$ f(x) = x^3 - 4x^2 + 5x - 2 $

We will manually calculate the gradient descent steps for a few iterations starting from an initial value of $ x = 2 $, using a fixed learning rate of $ \alpha = 0.1 $.

1. **Initial Value**: $ x = 2 $
2. **Learning Rate**: $ \alpha = 0.1 $
3. **Objective Function**: $ f(x) = x^3 - 4x^2 + 5x - 2 $
4. **Gradient**: $ \frac{df}{dx} = 3x^2 - 8x + 5 $

Let's compute the first few iterations manually:

**Iteration 1**:
- Current Value of $ x $: $ x = 2 $
- Gradient at $ x $: $ \frac{df}{dx} = 3(2)^2 - 8(2) + 5 = 3(4) - 16 + 5 = 12 - 16 + 5 = 1 $
- Update Rule: $ x_{\text{new}} = x - \alpha \frac{df}{dx} $
- New Value of $ x $: $ x_{\text{new}} = 2 - 0.1(1) = 2 - 0.1 = 1.9 $

**Iteration 2**:
- Current Value of $ x $: $ x = 1.9 $
- Gradient at $ x $: $ \frac{df}{dx} = 3(1.9)^2 - 8(1.9) + 5 = 3(3.61) - 15.2 + 5 = 10.83 - 15.2 + 5 = 0.63 $
- New Value of $ x $: $ x_{\text{new}} = 1.9 - 0.1(0.63) = 1.9 - 0.063 = 1.837 $

**Iteration 3**:
- Current Value of $ x $: $ x = 1.837 $
- Gradient at $ x $: $ \frac{df}{dx} = 3(1.837)^2 - 8(1.837) + 5 = 3(3.374569) - 14.696 + 5 = 10.123707 - 14.696 + 5 = 0.427707 $
- New Value of $ x $: $ x_{\text{new}} = 1.837 - 0.1(0.427707) = 1.837 - 0.042771 = 1.7942293 $

We can continue this process for more iterations to find the minimum of the function $ f(x) $ using gradient descent.

## Python visualization:

````python
import numpy as np
import matplotlib.pyplot as plt

# Define the objective function f(x)
def f(x):
    return x**3 - 4*x**2 + 5*x - 2

# Define the gradient of f(x)
def gradient(x):
    return 3*x**2 - 8*x + 5

# RMSProp algorithm
def rmsprop(x_initial, alpha=0.1, rho=0.9, epsilon=1e-8, num_iterations=100):
    x_values = [x_initial]  # Store the values of x at each iteration
    x = x_initial  # Initialize x
    cache = 0  # Initialize the cache
    
    for _ in range(num_iterations):
        grad = gradient(x)  # Compute the gradient at current x
        cache = rho * cache + (1 - rho) * grad**2  # Update the cache
        x = x - alpha * grad / (np.sqrt(cache) + epsilon)  # Update x using RMSProp
        x_values.append(x)  # Append the new x value
        
    return x_values

# Generate x values for plotting the function
x_vals = np.linspace(-1, 4, 400)
# Compute corresponding y values for the function
y_vals = f(x_vals)

# Apply RMSProp to minimize the function starting from an initial point
initial_point =4
x_rmsprop = rmsprop(initial_point)

# Plot the function
plt.figure(figsize=(10, 6))
plt.plot(x_vals, y_vals, label='f(x) = x^3 - 4x^2 + 5x - 2', color='blue')
# Plot RMSProp updates
plt.scatter(x_rmsprop, [f(x) for x in x_rmsprop], color='red', label='RMSProp Updates', marker='x')
plt.title('RMSProp Optimization')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)

# Set x and y axis limits based on the data points
plt.xlim(min(x_vals), max(x_vals))
plt.ylim(min(y_vals), max(y_vals))

plt.show()
````

# AADAM

The Adam algorithm is a popular optimization algorithm commonly used in machine learning. It combines concepts from AdaGrad, RMSProp, and momentum-based methods to efficiently update the parameters during training. Here's a breakdown of its key components:

1. **Exponentially Averaged Gradients**: Adam maintains an exponentially averaged value (denoted as Ai) for each parameter's gradient. This is similar to RMSProp, and it helps to normalize the gradients and reduce the impact of noisy updates.

2. **Exponentially Smoothed Gradient**: In addition to the averaged gradients, Adam also maintains an exponentially smoothed value (denoted as Fi) of the gradients themselves. This smoothing is performed using a different decay parameter (f) compared to Ai. It's similar to momentum-based methods and helps to incorporate momentum into the updates.

3. **Learning Rate Adaptation**: The learning rate () in Adam is adapted based on the iteration index (t). It's adjusted using a bias correction factor to account for the initialization bias of the averaged gradients and smoothed gradients. This helps to improve the stability of the optimization process, especially in the early iterations.

4. **Update Rule**: The parameter updates in Adam are computed using both the averaged gradients (Ai) and the smoothed gradients (Fi). The update rule involves dividing the smoothed gradient by the square root of the averaged gradient, and then multiplying it by the learning rate.

5. **Initialization Bias Correction**: Adam addresses the initialization bias of the averaged and smoothed gradients by applying a bias correction factor to the learning rate. This ensures that the optimization process remains stable and effective, even in the early iterations.

Overall, Adam combines the benefits of gradient normalization, momentum, and adaptive learning rates to achieve efficient and effective parameter updates during training. It's widely used in practice and often performs competitively with other optimization algorithms.

## Mathematical example:
Sure, let's consider a complex equation and go through a few steps of the Adam optimization algorithm manually. We'll choose the following equation:

$ f(x) = x^4 - 3x^3 + 2 $

Here's how we'll proceed:

1. Initialize the parameters:
   - Learning rate ($\alpha$)
   - Decay rates for exponentially averaged gradients ($\rho$) and smoothed gradients ($\rho_f$)
   - Initial values for the exponentially averaged gradients ($A$) and smoothed gradients ($F$)
   - Initial value for the parameter ($x$)

2. Compute the gradient of the function with respect to the parameter ($\frac{df}{dx}$).

3. Update the exponentially averaged gradients ($A$) and smoothed gradients ($F$) using the decay rates ($\rho$ and $\rho_f$).

4. Compute the bias-corrected learning rate ($\alpha_t$).

5. Update the parameter ($x$) using the exponentially averaged gradients ($A$), smoothed gradients ($F$), and bias-corrected learning rate ($\alpha_t$).

6. Repeat steps 2-5 for a few iterations.

Let's go through these steps manually for the given equation. We'll choose arbitrary initial values and parameters for demonstration purposes.

Let's go through a few steps of the Adam optimization algorithm manually for the given equation $ f(x) = x^4 - 3x^3 + 2 $.

Step 1: Initialization
- Learning rate ($\alpha$): 0.1
- Decay rates for exponentially averaged gradients ($\rho$): 0.9
- Decay rates for exponentially smoothed gradients ($\rho_f$): 0.99
- Initial value for the parameter ($x$): 2
- Initial values for exponentially averaged gradients ($A$): 0
- Initial values for exponentially smoothed gradients ($F$): 0

Step 2: Compute Gradient
$ \frac{df}{dx} = 4x^3 - 9x^2 $

At $ x = 2 $:
$ \frac{df}{dx} = 4(2)^3 - 9(2)^2 = 32 - 36 = -4 $

Step 3: Update Gradients

$ A \leftarrow \rho A + (1 - \rho) \frac{df}{dx} $

$ F \leftarrow \rho_f F + (1 - \rho_f) \left(\frac{df}{dx}\right)^2 $

At iteration 1:

$ A \leftarrow 0.9 \times 0 + (1 - 0.9) \times (-4) = -0.4 $

$ F \leftarrow 0.99 \times 0 + (1 - 0.99) \times (-4)^2 = 0.16 $

Step 4: Compute Bias-Corrected Learning Rate

$ \alpha_t = \alpha \times \frac{1}{1 - \rho^t} $

At iteration 1:

$ \alpha_t = 0.1 \times \frac{1}{1 - 0.9^1} = 0.1 \times \frac{1}{0.1} = 1 $

Step 5: Update Parameter

$ x \leftarrow x - \alpha_t \times \frac{A}{\sqrt{F} + \epsilon} $

At iteration 1:
$ x \leftarrow 2 - 1 \times \frac{-0.4}{\sqrt{0.16} + 10^{-8}} $

$ x \leftarrow 2 - 1 \times \frac{-0.4}{0.4 + 10^{-8}} $

$ x \leftarrow 2 + 0.4 = 2.4 $

So, after the first iteration, the parameter $ x $ is updated to 2.4.

We can continue these steps for further iterations to optimize the parameter $ x $ using the Adam algorithm.

## Python Visualization:

````python
import numpy as np
import matplotlib.pyplot as plt

# Define the function
def f(x):
    return x**3 - 2*x**2 + 2*x + 1

# Define the derivative of the function
def df(x):
    return 3*x**2 - 4*x + 2

# Adam optimization algorithm
def adam_optimization(x_init, alpha, rho, rho_f, epsilon, num_iterations):
    x = x_init
    A = 0
    F = 0
    x_history = [x]

    for i in range(num_iterations):
        grad = df(x)
        A = rho * A + (1 - rho) * grad
        F = rho_f * F + (1 - rho_f) * grad**2
        alpha_t = alpha * (1 / (1 - rho**(i+1)))
        x -= alpha_t * (A / (np.sqrt(F) + epsilon))
        x_history.append(x)
    
    return x_history

# Parameters for Adam
x_init = 2
alpha = 0.1
rho = 0.9
rho_f = 0.99
epsilon = 1e-8
num_iterations = 100

# Run Adam optimization
x_history = adam_optimization(x_init, alpha, rho, rho_f, epsilon, num_iterations)

# Plot the function and optimization trajectory
x_values = np.linspace(-1, 3, 400)
y_values = f(x_values)
plt.plot(x_values, y_values, label='f(x)')
plt.scatter(x_history, [f(x) for x in x_history], c='red', label='Optimization trajectory')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Adam Optimization')
plt.legend()
plt.grid(True)
plt.show()
````
